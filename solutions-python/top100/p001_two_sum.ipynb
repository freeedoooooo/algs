## 1. 题目一句话总结

给定一个整数数组和目标值，要找出两个下标，使得对应元素之和恰好等于目标值。

这题本质上是在考察：能不能用哈希表把“查找另一个数”从线性扫描优化成常数时间查询。

## 2. 题目理解

### 2.1 题目要求

数组中每个元素只能使用一次，需要返回那两个满足条件的元素下标。

题目保证一定存在唯一答案，所以我们不需要处理“无解”或“多解”的复杂分支。

### 2.2 输入与输出

- 输入：整数数组 `nums`，目标值 `target`
- 输出：两个下标组成的列表，例如 `[0, 1]`

### 2.3 关键约束

- 同一个元素不能重复使用
- 返回的是下标，不是元素值
- 题目保证存在且仅存在一个答案


## 3. 朴素思路

### 3.1 最直接的做法

使用两层循环，枚举每一对 `(i, j)`，判断 `nums[i] + nums[j]` 是否等于 `target`。

### 3.2 为什么不够好

- 时间复杂度是 `O(n^2)`
- 当数组变长时，重复比较会明显变多

### 3.3 关键观察

当我们遍历到当前数字 `x` 时，真正想知道的是：

```text
有没有一个之前出现过的数 = target - x
```

这说明问题可以转化成“快速查找补数是否出现过”，而这正是哈希表的强项。

## 4. 核心算法思路

### 4.1 算法名称

- 哈希表
- 一次遍历

### 4.2 为什么想到这个算法

如果当前数是 `num`，我们想找的另一个数就是：

```python
complement = target - num
```

所以我们可以：

1. 一边遍历数组
2. 一边把已经遍历过的数字存进哈希表
3. 每次先检查补数是否已经出现

### 4.3 关键步骤

1. 创建哈希表 `seen`，记录 `数字 -> 下标`
2. 遍历数组中的每个元素 `num`
3. 计算补数 `target - num`
4. 如果补数已经在 `seen` 里，直接返回答案
5. 否则把当前数字和下标存入 `seen`


## 5. 图解

以示例 `nums = [2, 7, 11, 15], target = 9` 为例。

### 5.1 初始状态

```text
seen = {}
```

### 5.2 遍历第 0 个数

当前数是 `2`，补数是 `9 - 2 = 7`。

```text
7 不在 seen 中
把 2 记进去：seen = {2: 0}
```

### 5.3 遍历第 1 个数

当前数是 `7`，补数是 `9 - 7 = 2`。

```text
2 在 seen 中，且对应下标是 0
所以答案就是 [0, 1]
```

整个过程可以写成：

```text
i = 0, num = 2, complement = 7, seen = {}
=> 未命中，记录 2:0

i = 1, num = 7, complement = 2, seen = {2:0}
=> 命中补数 2，返回 [0, 1]
```

这题的核心不是把所有两数组合都枚举出来，而是在遍历到当前位置时，快速知道“我要找的人是不是已经出现过了”。

````python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        seen: dict[int, int] = {}

        for i, num in enumerate(nums):
            complement = target - num

            if complement in seen:
                return [seen[complement], i]

            seen[num] = i

        raise ValueError('Input does not contain exactly one solution.')
````

## 6. 代码逐段讲解

### 6.1 哈希表里存什么

`seen` 记录的是“某个数字之前出现在哪个下标”。

### 6.2 为什么先查再存

因为题目不允许同一个元素使用两次，所以必须先检查补数是否已经在前面出现，再把当前元素放进去。

### 6.3 为什么能直接返回

题目保证答案唯一，因此第一次命中补数时就已经是正确答案。

## 7. 复杂度分析

- 时间复杂度：`O(n)`
- 空间复杂度：`O(n)`

原因：

- 数组只遍历一次
- 哈希表最多存下所有元素


## 8. 易错点

1. 把返回值写成元素值而不是下标。
2. 先把当前元素存进哈希表，再查补数，可能错误地用到自己。
3. 忘了题目允许重复数字，比如 `[3, 3]` 这种情况也可能是答案。
4. 用双层循环能做出来，但没有体现这题真正想考的哈希优化思路。

## 9. 面试时怎么讲

可以这样说：

> 朴素做法是双层循环枚举两数之和，复杂度是 `O(n^2)`。  
> 但当我遍历到当前数时，我真正需要的是“目标值减去当前数”有没有在前面出现过。  
> 所以我用哈希表记录已经遍历过的数字和下标，每次先查补数，再记录当前数。  
> 这样就把时间复杂度优化到了 `O(n)`。

## 10. 实际应用场景

这题在现实中可以抽象成：在一批数据里快速找到一对满足某个组合条件的对象。

### 10.1 具体业务案例：优惠券配对校验

#### 业务背景

电商系统里有一批优惠项，每个优惠项都有金额，系统要快速判断是否存在两个优惠项组合后正好达到目标门槛。

#### 输入是什么

- 一个优惠金额数组
- 一个目标门槛值

#### 算法介入点

系统在遍历优惠项时，需要快速判断“还差多少”这个金额是否已经出现过。

#### 输出是什么

输出是一对满足条件的优惠项位置，后续可以据此拿到原始记录。

#### 它解决了什么问题

它解决的是：如何避免暴力枚举所有两两组合，而是通过哈希表快速命中目标配对。

````python
solver = Solution()

print(solver.twoSum([2, 7, 11, 15], 9))
print(solver.twoSum([3, 2, 4], 6))
print(solver.twoSum([3, 3], 6))
````
