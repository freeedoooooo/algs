这道题要求按前序遍历顺序返回二叉树节点值，也就是按照 `根 -> 左 -> 右` 的顺序访问整棵树。

这份 notebook 直接参考仓库里的 Java 原方案，把 Java 代码、Java 注释思路和 Python 转写整理到同一份笔记里。


## Java 原代码

```java
package leetcode.p144_binary_tree_preorder_traversal;

import java.util.ArrayList;
import java.util.List;
import java.util.Stack;

public class BinaryTreePreorderTraversal {

    public static class TreeNode {
        public int val;
        public TreeNode left;
        public TreeNode right;

        public TreeNode(int v) {
            val = v;
        }
    }

    public static List<Integer> preorderTraversal(TreeNode head) {
        List<Integer> result = new ArrayList<>();
        if (head == null) {
            return result;
        }

        Stack<TreeNode> stack = new Stack<>();
        stack.push(head);

        while (!stack.isEmpty()) {
            head = stack.pop();
            result.add(head.val);

            if (head.right != null) {
                stack.push(head.right);
            }

            if (head.left != null) {
                stack.push(head.left);
            }
        }

        return result;
    }
}
```


## Java 原方案解析

Java 版本用的是非递归前序遍历，核心工具是栈。

可以把原注释整理成下面这套流程：

1. 如果根节点为空，直接返回空列表。
2. 先把根节点压栈。
3. 每次弹出一个节点，就立刻把它加入答案，因为前序遍历的第一个访问动作就是先看根。
4. 先压入右孩子，再压入左孩子。
5. 因为栈是后进先出，所以左孩子会先弹出，从而保证整体访问顺序是 `根 -> 左 -> 右`。

这里最关键的不是用了栈，而是压栈顺序。

- 如果先压左，再压右，弹出来会变成 `根 -> 右 -> 左`
- 只有先压右，再压左，弹栈时才会先处理左子树


## 图解这个栈过程

假设树是：

```text
    1
   / \
  2   3
 / \   \
4   5   6
```

遍历过程可以这样理解：

```text
初始: stack = [1]

弹出 1, 记录 1
先压 3, 再压 2
stack = [3, 2]

弹出 2, 记录 2
先压 5, 再压 4
stack = [3, 5, 4]

弹出 4, 记录 4
stack = [3, 5]

弹出 5, 记录 5
stack = [3]

弹出 3, 记录 3
压入 6
stack = [6]

弹出 6, 记录 6
stack = []
```

最终结果就是：`[1, 2, 4, 5, 3, 6]`。


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class TreeNode:
    val: int
    left: Optional["TreeNode"] = None
    right: Optional["TreeNode"] = None


class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> list[int]:
        # 对应 Java 里的 result 列表。
        result: list[int] = []
        if root is None:
            return result

        # 对应 Java 里的 Stack<TreeNode> stack。
        stack = [root]

        while stack:
            node = stack.pop()
            # 前序遍历是先访问根节点，所以弹出后立刻加入答案。
            result.append(node.val)

            # 先压右孩子，再压左孩子，这样左子树会先弹出。
            if node.right is not None:
                stack.append(node.right)
            if node.left is not None:
                stack.append(node.left)

        return result
````

## Python 转写时和 Java 的一一对应

- Java 的 `List<Integer> result` 对应 Python 的 `list[int]`
- Java 的 `Stack<TreeNode>` 对应 Python 的普通列表 `stack`
- Java 的 `stack.push(...)` / `stack.pop()` 对应 Python 的 `append(...)` / `pop()`
- Java 版本里的变量 `head` 在 Python 里改成了 `node`，语义更清晰，但流程完全一致

也就是说，这里不是换了一种思路，只是把同一套栈遍历逻辑翻译成 Python。


## 复杂度

- 时间复杂度：`O(n)`
- 空间复杂度：`O(n)`

每个节点都会入栈和出栈一次，所以总时间是线性的。辅助空间主要来自栈，最坏情况下可能达到 `O(n)`。


## 易错点

1. 把压栈顺序写反，导致结果变成 `根 -> 右 -> 左`。
2. 忘了处理空树，直接把 `None` 压栈后访问属性会报错。
3. 误以为前序遍历一定要先一路向左，那其实是另一种写法；当前 Java 原方案是弹栈即访问。
4. 用 Python 列表模拟栈时，若误用 `pop(0)`，就会把结构变成队列，顺序和复杂度都会不对。


````python
# 示例树:
#     1
#    / \
#   2   3
#  / \   \
# 4   5   6

root = TreeNode(
    1,
    TreeNode(2, TreeNode(4), TreeNode(5)),
    TreeNode(3, None, TreeNode(6)),
)

print(Solution().preorderTraversal(root))
print(Solution().preorderTraversal(None))
````

## 一句话复盘

> 这题的关键不是单纯记住前序遍历，而是记住用栈实现时必须先压右孩子、再压左孩子，才能保持 `根 -> 左 -> 右`。
