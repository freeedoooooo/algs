## 1. 题目一句话总结

这道题让我们把两个用逆序链表表示的整数逐位相加，并把结果也用同样的逆序链表形式返回。

本质上考的是链表上的竖式加法模拟，以及如何优雅地处理进位和结果链表构造。


## 2. 题目理解与约束分析

### 2.1 题目要求

两个非负整数不是直接给数字，而是拆成链表存储。每个节点只放一位，并且是逆序存储。

例如数字 `342` 会表示成链表：`2 -> 4 -> 3`。

我们需要返回它们相加之后的新链表。

### 2.2 输入与输出

- 输入：两个链表头节点 `l1`、`l2`
- 输出：相加结果链表的头节点
- 返回结果含义：返回的链表仍然采用逆序存储

### 2.3 关键约束

- 每个节点只存一位数字
- 两条链表长度可能不同
- 最后可能还会多出一位进位
- 结果通常需要构造一条新链表

### 2.4 示例分析

输入：`l1 = [2,4,3]`，`l2 = [5,6,4]`

它们表示的数字分别是：

- `342`
- `465`

逐位相加：

- `2 + 5 = 7`
- `4 + 6 = 10`，当前位写 `0`，进位 `1`
- `3 + 4 + 1 = 8`

所以结果是 `[7,0,8]`，表示数字 `807`。


## 3. Java 原代码

```java
package leetcode.p002_add_two_numbers;

public class AddTwoNumbers {

    public static class ListNode {
        public int val;
        public ListNode next;

        public ListNode(int val) {
            this.val = val;
        }

        public ListNode(int val, ListNode next) {
            this.val = val;
            this.next = next;
        }
    }

    public static ListNode addTwoNumbers(ListNode l1, ListNode l2) {
        if (l1 == null) {
            return l2;
        }
        if (l2 == null) {
            return l1;
        }

        ListNode head = new ListNode(0);
        ListNode current = head;
        int carry = 0;

        while (l1 != null || l2 != null) {
            int value1 = (l1 == null) ? 0 : l1.val;
            int value2 = (l2 == null) ? 0 : l2.val;

            int sum = value1 + value2 + carry;
            carry = sum / 10;
            int value = sum % 10;

            current.next = new ListNode(value);
            current = current.next;

            if (l1 != null) {
                l1 = l1.next;
            }
            if (l2 != null) {
                l2 = l2.next;
            }
        }

        if (carry != 0) {
            current.next = new ListNode(carry);
        }

        return head.next;
    }
}
```


## 4. 先从 Java 原方案理解

Java 原方案的核心其实非常朴素，就是模拟我们手算加法时的“逐位相加 + 进位”。

原代码里的关键点有这几个：

1. 如果某一条链表先走完，就把这一位当成 `0`。
2. 用一个虚拟头节点 `head` 来统一构造结果链表。
3. 每一轮都计算 `value1 + value2 + carry`。
4. 当前位保留 `sum % 10`，进位更新为 `sum / 10`。
5. 全部节点处理完以后，如果 `carry != 0`，还要额外补一个节点。

这个方案的优点是逻辑非常稳定，不需要先把链表转成整数，也不会遇到大整数溢出问题。


## 5. 朴素思路到优化思路

### 5.1 最直接的想法

一个很容易想到但不推荐的思路是：

1. 先把链表还原成整数
2. 做整数加法
3. 再把结果重新拆回链表

### 5.2 为什么不够好

- 链表可能很长，整数可能超出语言内置数值范围
- 先转整数再转回链表，做了很多题目并不需要的工作
- 题目本身已经把数字拆成“按位结构”，最自然的解法就是按位处理

### 5.3 优化方向

既然链表本来就是逆序存储，那链表头部刚好就是个位、十位、百位……

这意味着我们不需要翻转，也不需要先转整数，直接同步遍历两条链表，边走边算就行。


## 6. 核心算法讲解

### 6.1 算法名称

- 链表模拟
- 竖式加法

### 6.2 为什么想到这个算法

因为题目给出的存储方式本身就非常适合逐位运算：

- 链表当前节点就是当前位
- 两条链表同步往后走，就相当于从低位往高位计算
- 额外维护一个 `carry`，就能完整模拟竖式加法

### 6.3 关键状态

- `l1`：第一条链表当前处理位置
- `l2`：第二条链表当前处理位置
- `carry`：上一位产生的进位
- `dummy`：结果链表的虚拟头节点
- `cur`：结果链表的尾指针

### 6.4 步骤拆解

1. 创建虚拟头节点 `dummy` 和结果尾指针 `cur`
2. 初始化进位 `carry = 0`
3. 循环处理两条链表当前位，如果某条链表为空就把值视为 `0`
4. 计算总和 `total = value1 + value2 + carry`
5. 当前位写入 `total % 10`，进位更新为 `total // 10`
6. 循环结束后，如果还有进位，再补一个新节点
7. 返回 `dummy.next`


## 7. 过程演示

以 `l1 = [9, 9, 9]`，`l2 = [1]` 为例：

```text
初始:
carry = 0

第 1 轮:
9 + 1 + 0 = 10
写入 0, carry = 1

第 2 轮:
9 + 0 + 1 = 10
写入 0, carry = 1

第 3 轮:
9 + 0 + 1 = 10
写入 0, carry = 1

链表都结束后:
carry = 1
还需要补一个节点 1

结果:
[0, 0, 0, 1]
```

这正对应十进制里的：`999 + 1 = 1000`。


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class ListNode:
    val: int = 0
    next: Optional["ListNode"] = None


def build_linked_list(values: list[int]) -> Optional[ListNode]:
    dummy = ListNode()
    cur = dummy
    for value in values:
        cur.next = ListNode(value)
        cur = cur.next
    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> list[int]:
    ans = []
    while head is not None:
        ans.append(head.val)
        head = head.next
    return ans
````

````python
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # 对应 Java 里的虚拟头节点 head。
        dummy = ListNode()
        cur = dummy
        carry = 0

        # 这里比 Java 多把“最后还有进位”的情况也直接写进循环条件。
        while l1 is not None or l2 is not None or carry != 0:
            value1 = l1.val if l1 is not None else 0
            value2 = l2.val if l2 is not None else 0

            total = value1 + value2 + carry
            carry, digit = divmod(total, 10)

            cur.next = ListNode(digit)
            cur = cur.next

            if l1 is not None:
                l1 = l1.next
            if l2 is not None:
                l2 = l2.next

        return dummy.next
````

## 8. 代码逐段讲解

### 8.1 初始化部分

`dummy` 的作用是统一处理结果链表构造过程，避免“第一个节点要不要特殊处理”的分支判断。

`carry` 用来保存上一位的进位。

### 8.2 主循环

每次循环都做同一件事：

- 读出 `l1` 当前位
- 读出 `l2` 当前位
- 加上上一轮进位 `carry`
- 计算当前位数字和新的进位
- 把当前位数字接到结果链表尾部

### 8.3 更新答案

`cur.next = ListNode(digit)` 是把当前位接到结果链表后面。

然后 `cur = cur.next`，把尾指针向后推进一位，准备下次追加。

### 8.4 返回结果

最终返回 `dummy.next`，因为 `dummy` 自己只是辅助节点，不属于真正答案。


## 9. 复杂度分析

- 时间复杂度：`O(max(m, n))`
- 为什么是这个复杂度：两条链表各自只会被遍历一次
- 空间复杂度：`O(1)`（不计结果链表）
- 为什么是这个复杂度：只使用了若干指针和进位变量，没有额外辅助结构


## 10. 易错点

1. 忘了处理两条链表长度不同的情况，导致短链表走完后空指针报错。
2. 忘了把 `carry` 加进当前位总和里。
3. 忘了循环结束后可能还需要补最后一个进位节点。
4. 不用虚拟头节点时，容易把结果链表第一个节点的创建逻辑写乱。


## 11. 面试时怎么讲

可以这样概括：

> 这题我会直接在链表上模拟十进制加法。因为链表本身就是逆序存储，所以头节点刚好是个位，可以从低位一路往高位算。
> 我会用一个虚拟头节点来构造结果链表，再维护一个 `carry` 表示进位。
> 每一轮取出两个当前位，如果某条链表已经结束，就把它看成 `0`，然后计算当前位和新的进位。
> 最后如果还有剩余进位，再补一个节点即可。
> 时间复杂度是 `O(max(m, n))`，额外空间复杂度是 `O(1)`。


## 12. 实际应用场景

这题对应的真实问题可以理解成：对很长的数字序列做逐位累加，并且不能依赖语言内置的大整数。

### 具体业务案例：超长数字字段累加

#### 业务背景

某些金融、账单或高精度计算系统里，数字长度可能远超普通整数类型上限。

#### 输入是什么

输入是按位拆开的数字序列，可能来自链式存储、分段存储，或者流式传输的数据块。

#### 算法介入点

系统要把两段超长数值相加，但不能直接转成普通整数类型。

#### 输出是什么

输出是新的按位结果序列，仍然可以继续交给后续模块处理。

#### 价值是什么

它解决的是：在数值很长、不能整体装入普通整数类型时，如何稳定地完成加法运算。


````python
l1 = build_linked_list([2, 4, 3])
l2 = build_linked_list([5, 6, 4])
result = Solution().addTwoNumbers(l1, l2)
print('示例一:', linked_list_to_list(result))

l1 = build_linked_list([9, 9, 9])
l2 = build_linked_list([1])
result = Solution().addTwoNumbers(l1, l2)
print('进位示例:', linked_list_to_list(result))
````
