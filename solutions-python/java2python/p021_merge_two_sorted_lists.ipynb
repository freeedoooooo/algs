## 1. 题目一句话总结

这道题要求把两条已经递增有序的链表合并成一条新的递增链表。

本质上考的是链表上的有序归并，也就是归并排序里最核心的那一步。


## 2. 题目理解与约束分析

### 2.1 题目要求

给两条已经排好序的单链表，我们需要把它们合并成一条仍然有序的新链表。

这里的“新链表”不一定要求新建所有节点，更自然的做法是直接复用原来的节点，把它们重新接起来。

### 2.2 输入与输出

- 输入：两条递增链表的头节点 `list1`、`list2`
- 输出：合并后链表的头节点
- 返回结果含义：返回的链表需要保持整体递增

### 2.3 关键约束

- 两条链表都可能为空
- 两条链表长度可能不同
- 节点值允许重复
- 最好复用原链表节点，而不是额外创建整条新链表

### 2.4 示例分析

输入：`list1 = [1,2,4]`，`list2 = [1,3,4]`

比较过程是：

- 先比较两个 `1`，取其中一个接到结果链表
- 再比较 `2` 和另一个 `1`，取 `1`
- 再比较 `2` 和 `3`，取 `2`
- 再比较 `4` 和 `3`，取 `3`
- 再比较 `4` 和 `4`，依次接上

最终结果是：`[1,1,2,3,4,4]`。


## 3. Java 原代码

```java
package leetcode.p021_merge_two_sorted_lists;

import util.collection.ListNode;

public class MergeTwoSortedLists {

    public static ListNode mergeTwoListsRecursive(ListNode head1, ListNode head2) {
        if (head1 == null) {
            return head2;
        }
        if (head2 == null) {
            return head1;
        }
        if (head1.val < head2.val) {
            head1.next = mergeTwoListsRecursive(head1.next, head2);
            return head1;
        } else {
            head2.next = mergeTwoListsRecursive(head1, head2.next);
            return head2;
        }
    }

    public static ListNode mergeTwoListsIteration(ListNode head1, ListNode head2) {
        if (head1 == null) {
            return head2;
        }
        if (head2 == null) {
            return head1;
        }

        ListNode head = head1.val < head2.val ? head1 : head2;
        ListNode cur1 = head.next;
        ListNode cur2 = head == head1 ? head2 : head1;
        ListNode pre = head;

        while (cur1 != null && cur2 != null) {
            if (cur1.val < cur2.val) {
                pre.next = cur1;
                cur1 = cur1.next;
            } else {
                pre.next = cur2;
                cur2 = cur2.next;
            }
            pre = pre.next;
        }

        pre.next = cur1 != null ? cur1 : cur2;
        return head;
    }
}
```


## 4. 先从 Java 原方案理解

Java 原文件给了两种写法：

1. 递归版
2. 迭代版

虽然写法不同，但核心思路是一致的：

- 每次只比较两条链表当前节点值
- 把较小的那个节点接到结果链表后面
- 对应那条链表继续向后推进
- 某一边先结束后，把另一边剩余部分直接挂上去

这其实就是“两个有序序列的归并过程”，只不过数组版换成了链表版。


## 5. 朴素思路到优化思路

### 5.1 最直接的想法

一个比较笨的做法是：

1. 把两条链表所有值先取出来
2. 放进数组排序
3. 再重新建一条链表

### 5.2 为什么不够好

- 明明输入已经有序，却浪费了这个性质
- 额外引入数组和排序，复杂度更高
- 还会平白创建很多新节点

### 5.3 优化方向

既然两条链表本来就有序，那最自然的策略就是像归并排序那样，两个指针同时往前走，随时接出较小值。

这样可以做到：

- 时间复杂度 `O(m + n)`
- 迭代版额外空间 `O(1)`
- 尽量复用原节点


## 6. 核心算法讲解

### 6.1 算法名称

- 有序链表归并
- 双指针

### 6.2 为什么想到这个算法

因为两条链表都已经递增有序，所以结果链表的下一个节点一定只可能来自当前两个头部节点中较小的那个。

这说明我们完全没必要回头，也不需要全局排序，局部最优选择就能组成全局正确结果。

### 6.3 关键状态

在 Java 迭代版里，状态定义很清楚：

- `head`：最终结果链表的头节点
- `cur1`：当前正在比较的第一条链表节点
- `cur2`：当前正在比较的第二条链表节点
- `pre`：结果链表当前尾部节点

在 Python 里我会保留递归版，同时把主讲实现写成更容易理解的虚拟头节点迭代版。

### 6.4 步骤拆解

1. 处理空链表边界
2. 创建虚拟头节点 `dummy` 和尾指针 `tail`
3. 当两条链表都未结束时，比较当前值
4. 把较小节点接到 `tail.next`
5. 移动被选中的那条链表指针
6. 移动 `tail`
7. 循环结束后，把剩余链表整体接上
8. 返回 `dummy.next`


## 7. 过程演示

以 `list1 = [1,2,4]`，`list2 = [1,3,4]` 为例：

```text
初始:
list1: 1 -> 2 -> 4
list2: 1 -> 3 -> 4
result: dummy

第 1 轮:
比较 1 和 1，取右边这个 1
result: 1

第 2 轮:
比较 1 和 3，取左边这个 1
result: 1 -> 1

第 3 轮:
比较 2 和 3，取 2
result: 1 -> 1 -> 2

第 4 轮:
比较 4 和 3，取 3
result: 1 -> 1 -> 2 -> 3

第 5 轮:
比较 4 和 4，取右边这个 4
result: 1 -> 1 -> 2 -> 3 -> 4

结束:
一边走完后，把另一边剩余的 4 直接接上
result: 1 -> 1 -> 2 -> 3 -> 4 -> 4
```


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class ListNode:
    val: int = 0
    next: Optional["ListNode"] = None


def build_linked_list(values: list[int]) -> Optional[ListNode]:
    dummy = ListNode()
    cur = dummy
    for value in values:
        cur.next = ListNode(value)
        cur = cur.next
    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> list[int]:
    ans = []
    while head is not None:
        ans.append(head.val)
        head = head.next
    return ans
````

````python
class SolutionRecursive:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # 对应 Java 递归版的两个 base case。
        if list1 is None:
            return list2
        if list2 is None:
            return list1

        if list1.val < list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
````

````python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # 这里主推迭代版，因为更稳，也更适合面试现场展开。
        dummy = ListNode()
        tail = dummy

        while list1 is not None and list2 is not None:
            if list1.val <= list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next

        # 一边结束后，另一边剩余部分已经整体有序，直接接上即可。
        tail.next = list1 if list1 is not None else list2
        return dummy.next
````

## 8. 代码逐段讲解

### 8.1 递归版

递归版最像题目本身的定义：

- 当前较小节点就是结果链表头
- 它的 `next` 等于“剩余部分继续合并”的结果

写起来很优雅，但会消耗递归栈。

### 8.2 迭代版初始化

`dummy` 让我们统一处理链表拼接，不需要单独讨论第一个节点。

`tail` 永远指向当前结果链表最后一个节点。

### 8.3 主循环

每次只比较 `list1.val` 和 `list2.val`：

- 较小的节点接到 `tail.next`
- 对应链表向后走一步
- `tail` 也跟着前进

### 8.4 收尾部分

如果有一边先结束，另一边剩余节点本来就是递增有序的，因此可以整体挂到后面，不需要再逐个比较。


## 9. 复杂度分析

- 时间复杂度：`O(m + n)`
- 为什么是这个复杂度：两条链表里的节点最多都只会被访问一次
- 迭代版空间复杂度：`O(1)`
- 为什么是这个复杂度：除了少量指针变量，没有额外辅助结构
- 递归版空间复杂度：`O(m + n)` 的递归栈开销


## 10. 易错点

1. 忘了处理某一条链表本来就是空链表的情况。
2. 接上较小节点后，忘了移动对应链表指针。
3. 主循环结束后，忘了把剩余链表整体接上。
4. 递归版虽然短，但如果不理解“谁是当前头节点”，很容易把 `next` 接错。


## 11. 面试时怎么讲

可以这样概括：

> 这题我会按两个有序链表的归并过程来做。
> 因为两边都已经有序，所以每一轮结果链表的下一个节点，一定来自当前两个节点中较小的那个。

## 12. 实际应用场景

这题对应的真实问题可以理解成：把两路已经有序的数据流稳定合并成一路更大的有序流。

### 具体业务案例：合并两个时间有序事件流

#### 业务背景

一个系统可能同时从两个来源接收事件，比如两个服务节点各自产生日志或消息。

#### 输入是什么

输入是两路已经按时间排序好的事件序列。

#### 算法介入点

系统需要把两路数据合并，形成统一的全局有序输出。

#### 输出是什么

输出是一条仍然保持有序的事件流，供后续消费或存储。

#### 价值是什么

它解决的是：如何利用“输入已排序”这个条件，用线性时间完成稳定合并，而不是重新整体排序。


````python
list1 = build_linked_list([1, 2, 4])
list2 = build_linked_list([1, 3, 4])
result = Solution().mergeTwoLists(list1, list2)
print('示例一:', linked_list_to_list(result))

list1 = None
list2 = build_linked_list([0])
result = Solution().mergeTwoLists(list1, list2)
print('空链表边界:', linked_list_to_list(result))
````
