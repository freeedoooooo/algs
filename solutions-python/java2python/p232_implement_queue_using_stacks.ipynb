## 1. 题目一句话总结

这道题要求我们只用栈的合法操作去模拟一个先进先出的队列。

本质上考的是如何用后进先出的结构，通过两次翻转还原出先进先出的顺序。


## 2. 题目理解与约束分析

### 2.1 题目要求

实现一个 `MyQueue`，支持：

- `push(x)`
- `pop()`
- `peek()`
- `empty()`

但底层只能使用栈操作。

### 2.2 输入与输出

- 输入：一系列队列操作
- 输出：对应 `pop/peek/empty` 的结果
- 返回结果含义：行为必须与真实队列一致

### 2.3 关键约束

- 只能使用栈接口
- 队列要求先进先出
- 栈天然是后进先出，所以必须做顺序反转

### 2.4 示例分析

如果依次执行：

```text
push(1)
push(2)
peek() -> 1
pop() -> 1
empty() -> false
```

说明最早压入的元素必须最先被取出。


## 3. Java 原代码

```java
package leetcode.p232_implement_queue_using_stacks;

import java.util.Stack;

public class ImplementQueueUsingStacks {

    public static class MyQueue {
        private Stack<Integer> stack1;
        private Stack<Integer> stack2;

        public MyQueue() {
            stack1 = new Stack<>();
            stack2 = new Stack<>();
        }

        public void push(int x) {
            stack1.push(x);
        }

        public int pop() {
            if (empty()) {
                return -1;
            } else {
                while (!stack1.isEmpty()) {
                    stack2.push(stack1.pop());
                }
                int result = stack2.pop();
                while (!stack2.isEmpty()) {
                    stack1.push(stack2.pop());
                }
                return result;
            }
        }

        public int peek() {
            if (empty()) {
                return -1;
            } else {
                while (!stack1.isEmpty()) {
                    stack2.push(stack1.pop());
                }
                int result = stack2.peek();
                while (!stack2.isEmpty()) {
                    stack1.push(stack2.pop());
                }
                return result;
            }
        }

        public boolean empty() {
            return stack1.isEmpty();
        }
    }
}
```


## 4. 先从 Java 原方案理解

Java 原方案使用两个栈：

- `stack1`：正常接收新元素
- `stack2`：临时倒序辅助栈

它的思路很直接：

1. `push` 时只往 `stack1` 压
2. `pop/peek` 时把 `stack1` 所有元素倒到 `stack2`
3. 此时最早进入的元素就来到了 `stack2` 栈顶
4. 取完后再把元素倒回 `stack1`

这不是最省操作次数的写法，但它和 Java 原解完全一致，而且非常好理解。


## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

如果不限制结构，直接用队列就好了。

### 5.2 为什么这里难

因为栈的顺序是后进先出，而队列要先进先出。两次出入栈翻转，正好能把顺序倒回来。

### 5.3 Java 原方案的选择

更常见的优化版会把元素只在必要时从输入栈倒到输出栈，做到摊还 `O(1)`。但这里 notebook 故意保留 Java 原解的做法：每次 `pop/peek` 都完整倒过去再倒回来，这样更忠实于原题解。


## 6. 核心算法讲解

### 6.1 算法名称

- 双栈模拟队列
- 顺序翻转

### 6.2 为什么想到这个算法

因为一次栈翻转会把顺序反过来，两次翻转之后可以恢复相对顺序，同时把最早进入的元素放到可访问位置。

### 6.3 关键状态

- `stack1`：主栈，保存当前所有元素
- `stack2`：辅助栈，临时用于翻转顺序

### 6.4 步骤拆解

1. `push(x)`：直接压入 `stack1`
2. `pop()`：若为空返回 `-1`
3. 否则把 `stack1` 所有元素倒入 `stack2`
4. 从 `stack2` 取出或查看栈顶元素
5. 再把剩余元素倒回 `stack1`


## 7. 过程演示

依次执行：`push(1), push(2), push(3), pop()`

```text
初始：stack1 = [], stack2 = []

push(1)：stack1 = [1]
push(2)：stack1 = [1, 2]
push(3)：stack1 = [1, 2, 3]

执行 pop：
倒入 stack2 后：stack2 = [3, 2, 1]
stack2 栈顶就是 1
弹出 1
再把剩余元素倒回 stack1：stack1 = [2, 3]
```

这样最早入队的元素 `1` 被最先弹出，符合队列语义。


````python
class MyQueue:
    def __init__(self) -> None:
        self.stack1: list[int] = []
        self.stack2: list[int] = []

    def push(self, x: int) -> None:
        self.stack1.append(x)

    def pop(self) -> int:
        if self.empty():
            return -1

        while self.stack1:
            self.stack2.append(self.stack1.pop())

        result = self.stack2.pop()

        while self.stack2:
            self.stack1.append(self.stack2.pop())

        return result

    def peek(self) -> int:
        if self.empty():
            return -1

        while self.stack1:
            self.stack2.append(self.stack1.pop())

        result = self.stack2[-1]

        while self.stack2:
            self.stack1.append(self.stack2.pop())

        return result

    def empty(self) -> bool:
        return not self.stack1
````

## 8. 代码逐段讲解

### 8.1 `push`

`push` 很简单，直接压到 `stack1`。Java 原方案把所有重活都放在 `pop/peek`。

### 8.2 `pop`

`pop` 时先把 `stack1` 全部倒到 `stack2`，此时原来最早入队的元素就在 `stack2` 栈顶，可以直接弹出。

### 8.3 `peek`

`peek` 和 `pop` 基本同逻辑，只是不真正删除元素，而是查看后再完整倒回。

### 8.4 和更常见写法的区别

更常见的优化版会让 `stack2` 长期保留“输出顺序”，避免每次都全量倒回。但这里故意保留 Java 原方案，所以每次 `pop/peek` 都是完整搬运。


## 9. 复杂度分析

- `push()` 时间复杂度：`O(1)`
- `pop()`、`peek()` 时间复杂度：`O(n)`
- 为什么是这个时间复杂度：每次查询或弹出都要把元素倒过去再倒回来
- 空间复杂度：`O(n)`
- 为什么是这个空间复杂度：所有元素都存放在两个栈中


## 10. 易错点

1. 倒栈后忘记把元素再倒回去，导致后续状态错误。
2. `peek` 错误地把元素弹掉了，结果和 `pop` 混在一起。
3. 误以为本题 Java 原解一定是摊还 `O(1)` 版本，实际上它保留的是更直观但更慢的全量搬运版。
4. 空队列时没有单独返回约定值。


## 11. 面试时怎么讲

可以这样概括：

> 这题我先按 Java 原方案讲，会用两个栈模拟队列。
> `push` 时直接压入主栈；`pop/peek` 时把主栈元素全部倒到辅助栈，这样最早进入的元素就来到栈顶。
> 取值后再把元素倒回去，就能保持整体状态一致。
> 这版实现简单直接，代价是 `pop/peek` 为 `O(n)`。


## 12. 实际应用场景

这题可以类比成：底层只有 LIFO 结构，但业务层想伪装成 FIFO 通道。

### 具体业务案例：撤销栈适配消息队列语义

#### 业务背景

某些系统底层只暴露栈式缓存接口，但上层却希望按进入顺序消费请求。

#### 输入是什么

输入是一系列入队、出队、查看队头操作。

#### 算法介入点

系统通过双栈翻转，把栈顺序还原成队列顺序。

#### 输出是什么

输出是符合先进先出语义的处理结果。

#### 价值是什么

它解决的是底层结构受限但上层接口必须兼容的问题。


````python
queue = MyQueue()
queue.push(1)
queue.push(2)
queue.push(3)
print('当前队头:', queue.peek())
print('弹出:', queue.pop())
print('再次查看队头:', queue.peek())
print('是否为空:', queue.empty())
````

- 第一行应输出 `1`，说明最早入队元素能被正确看到。
- 第二行应输出 `1`，说明弹出顺序符合先进先出。
- 第三行应输出 `2`，说明剩余队头更新正确。
