## 1. 题目一句话总结

这道题要求我们找出两条单链表开始相交的第一个节点。

本质上考的是如何先判断是否真的相交，再把两条链表对齐到相同剩余长度后同步前进。


## 2. 题目理解与约束分析

### 2.1 题目要求

给两条单链表，返回它们开始相交的第一个公共节点。如果不相交，返回 `None`。

这里判断相交不是看节点值相同，而是看是否是同一个节点对象。

### 2.2 输入与输出

- 输入：两条链表头节点 `headA`、`headB`
- 输出：相交节点或 `None`
- 返回结果含义：返回两条链表首次共享的那个节点对象

### 2.3 关键约束

- 链表无环
- 不能修改原链表结构
- 节点值可以相同，但值相同不代表相交

### 2.4 示例分析

假设：

```text
A: 4 -> 1 -> 8 -> 4 -> 5
B: 5 -> 6 -> 1 -> 8 -> 4 -> 5
```

其中从值为 `8` 的节点开始，后面的链条完全是同一批节点对象，所以答案应该返回那个值为 `8` 的节点。


## 3. Java 原代码

```java
package leetcode.p160_intersection_of_two_linked_lists;

public class IntersectionOfTwoLinkedLists {

    public static class ListNode {
        public int val;
        public ListNode next;
    }

    public static ListNode getIntersectionNode(ListNode h1, ListNode h2) {
        if (h1 == null || h2 == null) {
            return null;
        }

        ListNode point1 = h1;
        ListNode point2 = h2;
        int diff = 0;

        while (point1.next != null) {
            point1 = point1.next;
            diff++;
        }
        while (point2.next != null) {
            point2 = point2.next;
            diff--;
        }

        if (point1 != point2) {
            return null;
        }

        if (diff >= 0) {
            point1 = h1;
            point2 = h2;
        } else {
            point1 = h2;
            point2 = h1;
        }
        diff = Math.abs(diff);

        while (diff-- != 0) {
            point1 = point1.next;
        }

        while (point1 != point2) {
            point1 = point1.next;
            point2 = point2.next;
        }

        return point1;
    }
}
```


## 4. 先从 Java 原方案理解

这份 Java 原方案不是常见的“双指针来回换头”模板，而是另一种非常清楚的写法：

1. 先分别走到两条链表尾部，并统计长度差
2. 如果尾节点不是同一个对象，直接判定不相交
3. 如果尾节点相同，说明一定相交
4. 让较长链表先走掉长度差
5. 然后两个指针同步前进，第一次相遇点就是答案

这个方案的优点是逻辑很强：它先用尾节点相等这个必要条件证明“确实相交”，再做后续定位。


## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

最直接的办法是用集合记录第一条链表所有节点，再去第二条链表里找第一个出现过的节点。

### 5.2 为什么还要优化

- 集合方案需要 `O(n)` 额外空间
- Java 原方案可以只用常数级额外空间

### 5.3 优化方向

两条无环链表如果相交，那么从相交点开始到尾部一定完全重合，因此尾节点必定相同。Java 原方案就是从这个性质切入。


## 6. 核心算法讲解

### 6.1 算法名称

- 长度差对齐
- 链表双指针

### 6.2 为什么想到这个算法

因为如果两条无环链表相交，那么它们尾部一定共享同一串节点，所以尾节点一定相同；同时，相交点之后的剩余长度一定相同。

只要先把较长链表多出来的那段走掉，两边就能在同一“剩余长度起点”上同步前进。

### 6.3 关键状态

- `point1`、`point2`：遍历链表的两个指针
- `diff`：两条链表长度差
- `h1`、`h2`：原始头节点

### 6.4 步骤拆解

1. 如果任一链表为空，直接返回 `None`
2. 分别走到两条链表尾部，同时统计长度差 `diff`
3. 如果尾节点不同，直接返回 `None`
4. 根据 `diff` 确定谁是长链表、谁是短链表
5. 让长链表先走 `abs(diff)` 步
6. 两个指针同步前进
7. 第一次相遇点就是相交点


## 7. 过程演示

假设：

```text
A: a1 -> a2 -> c1 -> c2
B: b1 -> c1 -> c2
```

### 7.1 先看尾节点

两条链表的尾节点都是 `c2`，说明它们一定相交。

### 7.2 算长度差

- A 长度更长，差值为 1

### 7.3 对齐后同步前进

让 A 先走 1 步：

```text
A: a2 -> c1 -> c2
B: b1 -> c1 -> c2
```

再同步走：

```text
a2 vs b1  不同
c1 vs c1  相同
```

所以第一个相交点就是 `c1`。


````python
from __future__ import annotations

from dataclasses import dataclass


@dataclass(eq=False)
class ListNode:
    val: int
    next: ListNode | None = None
````

````python
class Solution:
    def getIntersectionNode(self, h1: ListNode | None, h2: ListNode | None) -> ListNode | None:
        if h1 is None or h2 is None:
            return None

        point1 = h1
        point2 = h2
        diff = 0

        while point1.next is not None:
            point1 = point1.next
            diff += 1

        while point2.next is not None:
            point2 = point2.next
            diff -= 1

        if point1 is not point2:
            return None

        if diff >= 0:
            point1 = h1
            point2 = h2
        else:
            point1 = h2
            point2 = h1
        diff = abs(diff)

        while diff != 0:
            point1 = point1.next
            diff -= 1

        while point1 is not point2:
            point1 = point1.next
            point2 = point2.next

        return point1
````

## 8. 代码逐段讲解

### 8.1 第一步：比较尾节点

Java 原方案先走到两条链表的尾部。如果尾节点不是同一个对象，那就不可能相交，直接返回 `None`。

### 8.2 第二步：处理长度差

`diff` 为正，说明第一条链表更长；为负，说明第二条更长。然后让长链表先走掉这段差值。

### 8.3 第三步：同步前进

对齐后，两个指针距离尾部的剩余长度相同，所以第一次相遇位置就是相交起点。

### 8.4 为什么这里不改成别的常见模板

虽然这题还有“两个指针轮流换头”的经典写法，但这里特意保留 Java 原方案的长度差思路，这样 notebook 才真正是从 Java 到 Python 的迁移，而不是换了一套做法。


## 9. 复杂度分析

- 时间复杂度：`O(m + n)`
- 为什么是这个时间复杂度：两条链表各自线性遍历若干次，总量仍然是线性的
- 空间复杂度：`O(1)`
- 为什么是这个空间复杂度：只用了常数个指针变量


## 10. 易错点

1. 用节点值相等来判断相交，而不是用节点对象是否相同。
2. 忘记先比较尾节点，导致本来不相交的链表也继续做无意义对齐。
3. `diff` 为负时，没有交换长短链表的角色。
4. 对齐长度差后，没有再同步前进去找第一个相遇点。


## 11. 面试时怎么讲

可以这样概括：

> 我会先利用无环链表相交的性质：如果相交，尾节点一定相同。
> 所以我先分别走到两条链表尾部，并顺手统计长度差。
> 如果尾节点不同，直接返回空；如果尾节点相同，就让长链表先走掉长度差。
> 然后两个指针同步前进，第一次相遇的位置就是相交点。
> 这样时间复杂度 `O(m+n)`，空间复杂度 `O(1)`。


## 12. 实际应用场景

这题可以类比成：判断两条链式流程是否在某个节点之后合流。

### 具体业务案例：两条审批链是否汇合

#### 业务背景

某系统里两条处理链可能前半段不同，但在某个公共节点之后合并成统一流程。

#### 输入是什么

输入是两条链式流程定义，每个节点都通过 `next` 指向下一个环节。

#### 算法介入点

系统需要知道两条流程是否在某个节点开始共用后续链路。

#### 输出是什么

输出是开始汇合的第一个公共节点。

#### 价值是什么

它解决的是链式结构合流点定位问题，有助于流程复用分析和故障排查。


````python
common = ListNode(8, ListNode(4, ListNode(5)))
head_a = ListNode(4, ListNode(1, common))
head_b = ListNode(5, ListNode(6, ListNode(1, common)))

result = Solution().getIntersectionNode(head_a, head_b)
print('相交示例:', result.val if result else None)

head_c = ListNode(2, ListNode(6, ListNode(4)))
head_d = ListNode(1, ListNode(5))
result = Solution().getIntersectionNode(head_c, head_d)
print('不相交示例:', result)
````
