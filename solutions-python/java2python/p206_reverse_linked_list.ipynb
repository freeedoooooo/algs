## 1. 题目一句话总结

这道题要求我们把单链表原地反转，并返回新的头节点。

本质上考的是链表指针重连，而 Java 原文件里给出了迭代版、递归版，甚至还顺带写了双向链表反转。


## 2. 题目理解与约束分析

### 2.1 题目要求

给定一条单链表头节点 `head`，把整条链表反转后返回新的头节点。

### 2.2 输入与输出

- 输入：链表头节点 `head`
- 输出：反转后的新头节点
- 返回结果含义：原链表方向全部反过来

### 2.3 关键约束

- 链表可能为空
- 链表可能只有一个节点
- 希望原地完成，额外空间尽量少

### 2.4 示例分析

输入：`1 -> 2 -> 3 -> 4 -> 5`

输出：`5 -> 4 -> 3 -> 2 -> 1`

也就是说每个节点的 `next` 指针都要重新改向前一个节点。


## 3. Java 原代码

```java
package leetcode.p206_reverse_linked_list;

import util.collection.DoubleListNode;
import util.collection.ListNode;

public class ReverseLinkedList {

    public static ListNode reverseList(ListNode head) {
        if (head == null || head.next == null) {
            return head;
        }
        ListNode pre = null;
        ListNode next;
        while (head != null) {
            next = head.next;
            head.next = pre;
            pre = head;
            head = next;
        }
        return pre;
    }

    public static ListNode reverseListRecursive(ListNode head) {
        if (head == null || head.next == null) {
            return head;
        }

        ListNode newHead = reverseListRecursive(head.next);
        head.next.next = head;
        head.next = null;
        return newHead;
    }
}
```

Java 文件里还附带了 `reverseDoubleList`，说明这份原题解实际上是在系统整理链表反转的一整组操作。


## 4. 先从 Java 原方案理解

Java 原文件里主方法 `reverseList` 用的是最经典的迭代三指针写法：

- `pre`：当前节点反转后应该指向的前一个节点
- `head`：当前正在处理的节点
- `next`：提前保存原链表里的下一个节点

每一轮都做这三件事：

1. 先记住 `next = head.next`
2. 再把 `head.next` 改成 `pre`
3. 最后整体向前推进

Java 文件还提供了递归版 `reverseListRecursive`，它更优雅，但主实现显然更推荐迭代版，因为空间更稳。


## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

最朴素的想法可能是把节点值抄到数组里，再反着重建链表。

### 5.2 为什么不够好

- 多用了额外空间
- 没必要重建节点
- 题目本来就是想考指针反转

### 5.3 优化方向

既然链表结构本身已经在那，我们只要原地修改 `next` 指向即可。Java 原方案正是用三个指针一轮轮把方向倒过来。


## 6. 核心算法讲解

### 6.1 算法名称

- 链表原地反转
- 三指针迭代

### 6.2 为什么想到这个算法

因为每个节点只关心一件事：自己的 `next` 应该改成谁。只要在改向前先保存好旧的 `next`，就不会丢链。

### 6.3 关键状态

- `pre`：已经反转好的前半段头节点
- `head` / `cur`：当前处理节点
- `next_node`：原链表里的下一个节点

### 6.4 步骤拆解

1. `pre` 初始设为 `None`
2. 遍历链表，每次先保存 `next_node`
3. 把当前节点的 `next` 指向 `pre`
4. 让 `pre` 前进到当前节点
5. 让 `head` 前进到 `next_node`
6. 遍历结束后，`pre` 就是新头节点


## 7. 过程演示

以 `1 -> 2 -> 3` 为例：

```text
初始：
pre = None
head = 1 -> 2 -> 3

第一轮：
next = 2
1.next = None
pre = 1
head = 2

第二轮：
next = 3
2.next = 1
pre = 2 -> 1
head = 3

第三轮：
next = None
3.next = 2
pre = 3 -> 2 -> 1
head = None
```

结束后返回 `pre`，得到 `3 -> 2 -> 1`。


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class ListNode:
    val: int
    next: Optional["ListNode"] = None


def build_linked_list(values: list[int]) -> Optional[ListNode]:
    dummy = ListNode(0)
    cur = dummy
    for value in values:
        cur.next = ListNode(value)
        cur = cur.next
    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> list[int]:
    ans: list[int] = []
    while head is not None:
        ans.append(head.val)
        head = head.next
    return ans
````

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None or head.next is None:
            return head

        pre = None
        while head is not None:
            next_node = head.next
            head.next = pre
            pre = head
            head = next_node
        return pre

    def reverseListRecursive(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None or head.next is None:
            return head

        new_head = self.reverseListRecursive(head.next)
        head.next.next = head
        head.next = None
        return new_head
````

## 8. 代码逐段讲解

### 8.1 迭代版主逻辑

这里主实现严格对齐 Java 原解的迭代版。核心是“先存 next，再改 next”，顺序不能反。

### 8.2 为什么要先保存 `next_node`

因为一旦执行 `head.next = pre`，当前节点原本指向后续链表的那条边就被改掉了。如果没提前保存，下一个节点就丢了。

### 8.3 递归版补充

Java 文件里还给了递归版，所以 Python 这里也同步保留。递归版的关键句是：

```python
head.next.next = head
head.next = None
```

它表示：让后继节点反过来指向当前节点，同时把当前节点原来的 `next` 断掉。

### 8.4 为什么主讲迭代版

因为迭代版空间复杂度更稳，是 Java 原文件里最直接的核心实现。


## 9. 复杂度分析

- 迭代版时间复杂度：`O(n)`
- 为什么是这个时间复杂度：每个节点只访问一次
- 迭代版空间复杂度：`O(1)`
- 为什么是这个空间复杂度：只用了常数个指针变量
- 递归版空间复杂度：`O(n)`，主要来自递归调用栈


## 10. 易错点

1. 没先保存 `next` 就改 `head.next`，导致后续链表丢失。
2. 反转时把 `pre` 和 `head` 的更新顺序写错。
3. 递归版忘了写 `head.next = None`，会形成环。
4. 忘记处理空链表和单节点链表。


## 11. 面试时怎么讲

可以这样概括：

> 这题我会优先用迭代三指针做。
> 我维护 `pre` 表示已经反转好的前半段，`head` 表示当前节点，`next` 用来提前保存后续链表。
> 每一轮先存下一个节点，再把当前节点的 `next` 指向前一个节点，然后整体往前推进。
> 遍历结束后，`pre` 就是新的头节点。
> 时间复杂度 `O(n)`，额外空间复杂度 `O(1)`。


## 12. 实际应用场景

这题可以类比成：把一条单向流程链完全反向，改成从尾部往前处理。

### 具体业务案例：撤销操作链反向回放

#### 业务背景

某些系统会把操作历史组织成链表，后续希望从最新操作开始回滚。

#### 输入是什么

输入是一条按时间正向链接的操作链。

#### 算法介入点

系统需要把链表方向反过来，方便从最近操作开始逐步回退。

#### 输出是什么

输出是一条方向完全相反的新链表结构。

#### 价值是什么

它解决的是链式历史结构反向处理的问题。


````python
head = build_linked_list([1, 2, 3, 4, 5])
print('迭代版:', linked_list_to_list(Solution().reverseList(head)))

head = build_linked_list([1, 2, 3])
print('递归版:', linked_list_to_list(Solution().reverseListRecursive(head)))

head = build_linked_list([1])
print('单节点:', linked_list_to_list(Solution().reverseList(head)))
````

- 第二组应输出 `[3, 2, 1]`，说明递归版也和 Java 原文件的补充方案一致。
- 第三组应输出 `[1]`，说明边界情况下不会出错。
