## 1. 题目一句话总结

这道题要求我们按“左 -> 右 -> 根”的顺序返回二叉树节点值。

本质上考的是如何把后序遍历改写成非递归形式，而 Java 原方案给出的思路是非常经典的双栈写法。


## 2. 题目理解与约束分析

### 2.1 题目要求

给定一棵二叉树，返回它的后序遍历结果。

后序遍历固定顺序是：

- 先左子树
- 再右子树
- 最后根节点

### 2.2 输入与输出

- 输入：二叉树根节点 `root`
- 输出：整数列表 `list[int]`
- 返回结果含义：列表顺序就是后序遍历顺序

### 2.3 关键约束

- 树可能为空
- 可以递归，也可以迭代
- 这里要对照 Java 原方案，主讲双栈写法

### 2.4 示例分析

对于这棵树：

```text
    1
   / \
  2   3
 / \   \
4   5   6
```

后序遍历顺序是：

- 左子树：`4, 5, 2`
- 右子树：`6, 3`
- 根节点：`1`

所以结果是：`[4, 5, 2, 6, 3, 1]`


## 3. Java 原代码

```java
package leetcode.p145_binary_tree_postorder_traversal;

import java.util.ArrayList;
import java.util.List;
import java.util.Stack;

public class BinaryTreePostorderTraversal {

    public static class TreeNode {
        public int val;
        public TreeNode left;
        public TreeNode right;

        public TreeNode(int v) {
            val = v;
        }
    }

    public static List<Integer> postorderTraversal(TreeNode head) {
        List<Integer> result = new ArrayList<>();
        if (head == null) {
            return result;
        }

        Stack<TreeNode> stack = new Stack<>();
        Stack<TreeNode> collect = new Stack<>();
        stack.push(head);

        while (!stack.isEmpty()) {
            head = stack.pop();
            collect.push(head);

            if (head.left != null) {
                stack.push(head.left);
            }

            if (head.right != null) {
                stack.push(head.right);
            }
        }

        while (!collect.isEmpty()) {
            result.add(collect.pop().val);
        }

        return result;
    }
}
```


## 4. 先从 Java 原方案理解

Java 原方案使用了两个栈：

- `stack`：负责遍历
- `collect`：负责暂存反序结果

这套思路本质上是在借用一种“变形前序遍历”：

- 标准前序是 `根 -> 左 -> 右`
- Java 这里实际做的是 `根 -> 右 -> 左`
- 再把这个顺序反过来，就得到 `左 -> 右 -> 根`

所以关键不是死记双栈，而是看懂它在利用顺序反转。


## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

最直接的办法当然是递归：

1. 递归左子树
2. 递归右子树
3. 访问根节点

### 5.2 为什么还要看这个 Java 方案

- 面试里经常会追问非递归写法
- 双栈法逻辑清楚，比单栈后序更容易讲
- Java 原仓库这题就是主推双栈版本

### 5.3 优化方向

这里的“优化”不是复杂度上的巨大变化，而是把递归改写成显式栈，让遍历顺序可控、可解释。


## 6. 核心算法讲解

### 6.1 算法名称

- 二叉树后序遍历
- 双栈迭代

### 6.2 为什么想到这个算法

因为后序遍历 `左 -> 右 -> 根` 比较难直接“弹出即访问”。Java 原方案换了个角度：先生成 `根 -> 右 -> 左` 的访问顺序，再整体反转。

### 6.3 关键状态

- `stack`：主遍历栈
- `collect`：收集反序节点的辅助栈
- `result`：最终答案列表

### 6.4 步骤拆解

1. 根节点先入 `stack`
2. 循环弹出 `stack` 栈顶节点，并压入 `collect`
3. 先把左孩子压入 `stack`
4. 再把右孩子压入 `stack`
5. 这样 `collect` 里的顺序就相当于 `根 -> 右 -> 左`
6. 最后不断弹出 `collect`，得到 `左 -> 右 -> 根`


## 7. 过程演示

还是用这棵树：

```text
    1
   / \
  2   3
 / \   \
4   5   6
```

遍历阶段：

```text
初始：stack = [1], collect = []

弹 1，压入 collect
stack 压 2、3
collect = [1]

弹 3，压入 collect
stack 压 6
collect = [1, 3]

弹 6，压入 collect
collect = [1, 3, 6]

弹 2，压入 collect
stack 压 4、5
collect = [1, 3, 6, 2]

弹 5，压入 collect
collect = [1, 3, 6, 2, 5]

弹 4，压入 collect
collect = [1, 3, 6, 2, 5, 4]
```

最后把 `collect` 反向弹出，得到：`[4, 5, 2, 6, 3, 1]`


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class TreeNode:
    val: int
    left: Optional["TreeNode"] = None
    right: Optional["TreeNode"] = None
````

````python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> list[int]:
        result: list[int] = []
        if root is None:
            return result

        stack: list[TreeNode] = [root]
        collect: list[TreeNode] = []

        while stack:
            node = stack.pop()
            collect.append(node)

            # 对照 Java 原解：这里先压左、再压右。
            # 这样 collect 的弹出顺序才会变成后序遍历。
            if node.left is not None:
                stack.append(node.left)
            if node.right is not None:
                stack.append(node.right)

        while collect:
            result.append(collect.pop().val)

        return result
````

## 8. 代码逐段讲解

### 8.1 初始化部分

如果根节点为空，直接返回空列表；否则先把根节点放进 `stack`。

### 8.2 第一层栈遍历

每次从 `stack` 弹出一个节点后，不立即写入最终结果，而是先放进 `collect`。这相当于先构造一份“反过来的遍历顺序”。

### 8.3 为什么是先左后右

这一步非常关键。因为 `stack` 是后进先出：

- 先压左
- 再压右
- 弹出顺序就会变成 `根 -> 右 -> 左`

再经过 `collect` 的逆序弹出，最终才变成 `左 -> 右 -> 根`。

### 8.4 收尾部分

最后不断弹出 `collect`，就把临时反序结果还原成标准后序遍历顺序。


## 9. 复杂度分析

- 时间复杂度：`O(n)`
- 为什么是这个时间复杂度：每个节点最多进出两个栈各一次
- 空间复杂度：`O(n)`
- 为什么是这个空间复杂度：两个栈在最坏情况下都可能存储较多节点


## 10. 易错点

1. 把 `stack` 的压栈顺序写反，导致最终顺序不再是后序。
2. 误以为弹出 `stack` 时就能直接写入结果，其实 Java 原方案是先写到 `collect`。
3. 忘记处理空树，直接把 `None` 放进栈。
4. 把双栈和单栈后序遍历的思路混在一起，导致逻辑越写越乱。


## 11. 面试时怎么讲

可以这样概括：

> 这题我会按照 Java 原方案用双栈做。
> 第一层栈负责遍历节点，第二层栈负责把访问顺序暂存成反序。
> 具体来说，我先构造 `根 -> 右 -> 左` 的顺序，再把它反过来，正好得到 `左 -> 右 -> 根`。
> 这样写比单栈后序更直观，也更容易讲清楚。
> 时间复杂度 `O(n)`，空间复杂度 `O(n)`。


## 12. 实际应用场景

这题可以对应到一种很常见的场景：子任务必须先处理完，父任务才能收尾。

### 具体业务案例：文件夹大小汇总

#### 业务背景

在文件系统或对象存储里，一个目录节点通常依赖其所有子目录和文件的统计结果。

#### 输入是什么

输入是一棵目录树，父节点下面挂着多个子节点。

#### 算法介入点

系统要先汇总子节点的大小，再回到父节点做总计。

#### 输出是什么

输出是按“先子后父”顺序处理后的节点序列或聚合结果。

#### 价值是什么

它解决的是依赖方向由下往上的处理顺序问题，这和后序遍历的思想完全一致。


````python
root = TreeNode(
    1,
    TreeNode(2, TreeNode(4), TreeNode(5)),
    TreeNode(3, None, TreeNode(6)),
)

print('示例树:', Solution().postorderTraversal(root))
print('空树:', Solution().postorderTraversal(None))
````
