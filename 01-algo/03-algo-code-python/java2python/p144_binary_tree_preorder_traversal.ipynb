## 1. 题目一句话总结

这道题要求我们按“根 -> 左 -> 右”的顺序返回二叉树节点值。

本质上考的是前序遍历的访问顺序，以及如何用显式栈正确模拟这种顺序。


## 2. 题目理解与约束分析

### 2.1 题目要求

给定一棵二叉树，返回它的前序遍历结果。

前序遍历的访问顺序固定为：

- 先访问根节点
- 再访问左子树
- 最后访问右子树

### 2.2 输入与输出

- 输入：二叉树根节点 `root`
- 输出：整数列表 `list[int]`
- 返回结果含义：列表顺序就是前序遍历顺序

### 2.3 关键约束

- 树可能为空
- 可以用递归，也可以用迭代
- Java 原仓库这题用的是显式栈迭代版

### 2.4 示例分析

对于这棵树：

```text
    1
   / \
  2   3
 / \   \
4   5   6
```

前序遍历顺序是：

1. 先访问根 `1`
2. 再访问左子树 `2 -> 4 -> 5`
3. 最后访问右子树 `3 -> 6`

所以结果是：`[1, 2, 4, 5, 3, 6]`


## 3. Java 原代码

```java
package leetcode.p144_binary_tree_preorder_traversal;

import java.util.ArrayList;
import java.util.List;
import java.util.Stack;

public class BinaryTreePreorderTraversal {

    public static class TreeNode {
        public int val;
        public TreeNode left;
        public TreeNode right;

        public TreeNode(int v) {
            val = v;
        }
    }

    public static List<Integer> preorderTraversal(TreeNode head) {
        List<Integer> result = new ArrayList<>();
        if (head == null) {
            return result;
        }

        Stack<TreeNode> stack = new Stack<>();
        stack.push(head);

        while (!stack.isEmpty()) {
            head = stack.pop();
            result.add(head.val);

            if (head.right != null) {
                stack.push(head.right);
            }

            if (head.left != null) {
                stack.push(head.left);
            }
        }

        return result;
    }
}
```


## 4. 先从 Java 原方案理解

Java 原方案使用的是显式栈，而不是递归。

它的主流程非常简洁：

1. 根节点先入栈
2. 每次弹出一个节点，就立刻访问它
3. 然后先压右孩子，再压左孩子

这里真正的关键不是“用了栈”，而是“压栈顺序”。

因为栈是后进先出：

- 先压右，后压左
- 弹出时左孩子就会先出来
- 这样整体顺序才会保持为 `根 -> 左 -> 右`


## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

最容易想到的是递归：

1. 访问当前节点
2. 递归左子树
3. 递归右子树

### 5.2 为什么还要看迭代版

- 递归虽然短，但面试里经常会被追问非递归写法
- Java 原仓库本题就是显式栈版本
- 迭代版更能看清楚遍历顺序是如何被维护的

### 5.3 优化方向

这题的“优化”更多是实现方式上的：把递归的隐式调用栈改写成我们可控的显式栈。


## 6. 核心算法讲解

### 6.1 算法名称

- 二叉树前序遍历
- 显式栈迭代

### 6.2 为什么想到这个算法

因为前序遍历要求“访问根节点”的动作发生得最早，所以很适合用栈来做：每次从栈顶取出一个节点，立即访问，再把它的子节点按正确顺序压回去。

### 6.3 关键状态

- `stack`：保存待访问节点
- `result`：保存前序遍历结果
- `node`：当前弹出的节点

### 6.4 步骤拆解

1. 如果根节点为空，直接返回空列表
2. 否则先把根节点压栈
3. 循环弹出栈顶节点，立刻记录其值
4. 先把右孩子压栈
5. 再把左孩子压栈
6. 循环结束后返回结果列表


## 7. 过程演示

以这棵树为例：

```text
    1
   / \
  2   3
 / \   \
4   5   6
```

栈的变化过程是：

```text
初始：stack = [1]

弹出 1，记录 1
压 3，再压 2
stack = [3, 2]

弹出 2，记录 2
压 5，再压 4
stack = [3, 5, 4]

弹出 4，记录 4
stack = [3, 5]

弹出 5，记录 5
stack = [3]

弹出 3，记录 3
压 6
stack = [6]

弹出 6，记录 6
stack = []
```

最终结果就是：`[1, 2, 4, 5, 3, 6]`


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class TreeNode:
    val: int
    left: Optional["TreeNode"] = None
    right: Optional["TreeNode"] = None
````

````python
class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> list[int]:
        result: list[int] = []
        if root is None:
            return result

        stack: list[TreeNode] = [root]

        while stack:
            node = stack.pop()
            result.append(node.val)

            # 对照 Java 原解：一定要先压右，再压左。
            if node.right is not None:
                stack.append(node.right)
            if node.left is not None:
                stack.append(node.left)

        return result
````

## 8. 代码逐段讲解

### 8.1 初始化部分

如果 `root` 为空，直接返回空列表；否则把根节点先放进栈里，作为第一个待访问节点。

### 8.2 主循环

每次循环弹出一个节点后，立刻把它的值加入结果，因为前序遍历本来就是先访问根。

### 8.3 子节点入栈顺序

这一步是整题最关键的地方：

- 先压右孩子
- 再压左孩子

因为栈后进先出，所以左孩子会先弹出，从而保证整体顺序正确。

### 8.4 返回结果

当栈清空时，说明所有节点都已经被访问过，此时返回 `result`。


## 9. 复杂度分析

- 时间复杂度：`O(n)`
- 为什么是这个时间复杂度：每个节点最多入栈一次、出栈一次、访问一次
- 空间复杂度：`O(n)`
- 为什么是这个空间复杂度：辅助栈在最坏情况下可能保存较多节点；更精确地说，与树高和形态有关，最坏可到 `O(n)`


## 10. 易错点

1. 把压栈顺序写反了，导致结果变成 `根 -> 右 -> 左`。
2. 没处理空树，直接把 `None` 当节点访问。
3. 错把 `pop(0)` 当成栈弹出，实际上那会更像队列。
4. 误以为前序遍历必须写成“一路向左”的形式，其实 Java 原方案这种弹栈即访问也完全正确。


## 11. 面试时怎么讲

可以这样概括：

> 这题我会用显式栈做前序遍历。
> 做法是先把根节点压栈，然后每次弹出一个节点就立刻访问它。
> 为了保持前序的 `根 -> 左 -> 右` 顺序，我会先压右孩子，再压左孩子。
> 因为栈是后进先出，所以左子树会先被处理。
> 时间复杂度 `O(n)`，最坏空间复杂度 `O(n)`。


## 12. 实际应用场景

这题可以类比成：需要按“先父节点、再子节点”的顺序遍历一棵层级结构树。

#### 业务背景

输入是一棵组件树，每个节点代表一个组件或容器。

#### 输出是什么

输出是一个按前序遍历展开后的节点序列。

#### 价值是什么

它解决的是层级结构的顺序展开问题，特别适合“父节点必须先于子节点处理”的场景。


````python
root = TreeNode(
    1,
    TreeNode(2, TreeNode(4), TreeNode(5)),
    TreeNode(3, None, TreeNode(6)),
)
print('示例树:', Solution().preorderTraversal(root))
print('空树:', Solution().preorderTraversal(None))
````
