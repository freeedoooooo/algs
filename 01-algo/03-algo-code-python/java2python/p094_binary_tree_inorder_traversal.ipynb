## 1. 题目一句话总结

这道题要求我们按照“左子树 -> 根节点 -> 右子树”的顺序返回整棵二叉树的节点值。

本质上是在考察二叉树遍历顺序的理解，以及如何用显式栈把递归版中序遍历改写成迭代版。


## 2. 题目理解与约束分析

### 2.1 题目要求

给定一棵二叉树的根节点 `root`，返回它的中序遍历结果。

中序遍历的固定顺序是：

- 先遍历左子树
- 再访问当前节点
- 最后遍历右子树

### 2.2 输入与输出

- 输入：二叉树根节点 `root`
- 输出：整数列表 `list[int]`
- 返回结果含义：列表中元素顺序就是中序遍历顺序

### 2.3 关键约束

- 树可能为空
- 节点数量可能较多
- 递归和迭代都能做，但 Java 原方案用的是显式栈迭代

### 2.4 示例分析

输入：`root = [1,null,2,3]`

对应的树结构是：

```text
    1
     \
      2
     /
    3
```

按中序遍历：

- 先访问 `1` 的左子树，没有
- 访问 `1`
- 再到右子树 `2`
- 访问 `2` 前，先访问它的左子树 `3`

所以结果是：`[1, 3, 2]`


## 3. Java 原代码

```java
package leetcode.p094_binary_tree_inorder_traversal;

import java.util.ArrayList;
import java.util.List;
import java.util.Stack;

public class BinaryTreeInorderTraversal {

    public static class TreeNode {
        public int val;
        public TreeNode left;
        public TreeNode right;

        public TreeNode(int v) {
            val = v;
        }
    }

    public static List<Integer> inorderTraversal(TreeNode head) {
        List<Integer> result = new ArrayList<>();
        if (head != null) {
            Stack<TreeNode> stack = new Stack<>();
            while (!stack.isEmpty() || head != null) {
                if (head != null) {
                    stack.push(head);
                    head = head.left;
                } else {
                    head = stack.pop();
                    result.add(head.val);
                    head = head.right;
                }
            }
        }
        return result;
    }
}
```


## 4. 先从 Java 原方案理解

这份 Java 代码没有用递归，而是用 `Stack<TreeNode>` 显式模拟递归过程。

它的核心思想可以概括成一句话：

> 先一路往左走，把沿途节点都压栈；走到尽头后，再弹栈访问当前节点，然后转向右子树继续重复。

Java 方案里最关键的外层循环条件是：

```java
while (!stack.isEmpty() || head != null)
```

这表示：

- 只要当前节点还没走空，就继续往左压栈
- 或者虽然当前节点走空了，但栈里还存着待访问的祖先节点，也不能停

它和递归版中序遍历本质上一样，只是把系统调用栈换成了我们手动维护的栈。


## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

最自然的朴素做法其实是递归：

1. 递归遍历左子树
2. 记录根节点
3. 递归遍历右子树

### 5.2 为什么还要看迭代版

- 递归虽然短，但不容易把执行过程讲清楚
- 面试里经常会追问如何改成非递归
- Java 原仓库这题本来就主推显式栈方案

### 5.3 优化方向

所谓“优化”在这题里不主要是复杂度优化，而是实现方式优化：把隐式递归栈改成显式栈，让遍历过程更可控、更贴近底层机制。


## 6. 核心算法讲解

### 6.1 算法名称

- 二叉树中序遍历
- 显式栈迭代

### 6.2 为什么想到这个算法

中序遍历的访问顺序要求“左边全部处理完之后，才能访问当前节点”。这意味着我们需要一种机制，把“稍后再访问”的节点先记住。

栈正好满足这个需求：

- 往左走时，把节点压栈，表示它还不能立刻访问
- 左边走到底后，弹出最近那个节点，它此时正好轮到被访问

### 6.3 关键状态

- `stack`：保存还没访问值的节点路径
- `cur`：当前正在向下探索的节点
- `result`：保存最终中序结果

### 6.4 步骤拆解

1. 准备空栈 `stack` 和答案数组 `result`
2. 令 `cur = root`
3. 当 `cur` 不为空时，不断压栈并向左走
4. 当左边走到头后，弹出栈顶节点并记录其值
5. 把 `cur` 切到这个节点的右子树
6. 重复上面的过程，直到 `cur` 为空且栈也为空


## 7. 过程演示

还是用这棵树：

```text
    1
     \
      2
     /
    3
```

遍历过程可以写成：

```text
初始：
cur = 1, stack = [], result = []

步骤 1：一路向左
push 1
cur -> None

步骤 2：左边到底，弹栈访问
pop 1, result = [1]
cur -> 1.right = 2

步骤 3：继续一路向左
push 2
cur -> 3
push 3
cur -> None

步骤 4：弹栈访问 3
result = [1, 3]
cur -> 3.right = None

步骤 5：再弹栈访问 2
result = [1, 3, 2]
cur -> 2.right = None

结束：
cur = None, stack = []
```

这个过程说明：栈里存的并不是“访问过的节点”，而是“左边处理还没彻底完成、值还没加入答案的节点”。


````python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional


@dataclass
class TreeNode:
    val: int
    left: Optional["TreeNode"] = None
    right: Optional["TreeNode"] = None
````

````python
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> list[int]:
        # 对照 Java 原解：result 对应 ArrayList，stack 对应 Stack<TreeNode>。
        result: list[int] = []
        stack: list[TreeNode] = []
        cur = root

        while stack or cur is not None:
            if cur is not None:
                stack.append(cur)
                cur = cur.left
            else:
                cur = stack.pop()
                result.append(cur.val)
                cur = cur.right

        return result
````

## 8. 代码逐段讲解

### 8.1 初始化部分

`result` 用来收集答案，`stack` 用来手动模拟递归调用路径，`cur` 表示当前正在尝试往下走的节点。

### 8.2 主循环

循环里分成两个分支：

- 如果 `cur` 不为空：说明还能继续向左深入，就先压栈再走向 `cur.left`
- 如果 `cur` 为空：说明左边已经到底了，该弹出最近的祖先节点访问，然后转去它的右子树

### 8.3 更新答案

只有在 `cur = stack.pop()` 之后才会执行 `result.append(cur.val)`，这保证了左子树一定先于根节点被处理。

### 8.4 返回结果

当 `cur` 为空且栈也为空时，说明整棵树的所有节点都按中序访问完成，此时直接返回 `result`。


## 9. 复杂度分析

- 时间复杂度：`O(n)`
- 为什么是这个时间复杂度：每个节点最多入栈一次、出栈一次、访问一次
- 空间复杂度：`O(h)`
- 为什么是这个空间复杂度：栈里最多同时保存一条从根到叶的路径，`h` 是树高；最坏退化成链表时为 `O(n)`


## 10. 易错点

1. 把访问节点值的时机写早了，导致顺序变成前序或别的遍历顺序。
2. 只写 `while cur:`，漏掉“当前节点空了但栈里还有待处理节点”的情况。
3. 弹栈访问后忘了转向 `cur.right`，结果右子树完全没遍历。
4. 误以为栈里保存的是访问过的节点，其实它保存的是还没输出值的路径节点。


## 11. 面试时怎么讲

可以这样概括：

> 这题我会用显式栈做中序遍历，顺序是左、根、右。
> 做法是先一路往左，把沿途节点都压栈；走到头之后，弹出栈顶节点访问它的值，再转去它的右子树。
> 整个过程不断重复，直到当前节点为空且栈也为空。
> 这样每个节点只处理常数次，时间复杂度 `O(n)`，额外空间复杂度是 `O(h)`。


## 12. 实际应用场景

这题背后的真实能力是：按某种固定顺序遍历树形结构，并且在不使用递归时自己管理回溯路径。

### 具体业务案例：表达式语法树顺序读取

#### 业务背景

编译器或规则引擎里，经常会把表达式解析成一棵树，比如比较条件、算术表达式、布尔运算式。

#### 输入是什么

输入是一棵表达式树，节点可能是操作符，也可能是操作数。

#### 算法介入点

系统需要按固定顺序扫描整棵树，并在扫描时决定什么时候处理当前节点、什么时候继续深入子树。

#### 输出是什么

输出是一段按顺序展开后的节点序列，供后续优化、计算或展示模块使用。

#### 价值是什么

它解决的是树结构顺序访问与回溯管理问题，尤其适合在不希望依赖递归深度时使用。


````python
# 示例一：root = [1, null, 2, 3]
root = TreeNode(1, None, TreeNode(2, TreeNode(3), None))
print('示例一:', Solution().inorderTraversal(root))

# 示例二：空树
print('空树:', Solution().inorderTraversal(None))

# 示例三：完整小树
root = TreeNode(2, TreeNode(1), TreeNode(3))
print('平衡树:', Solution().inorderTraversal(root))
````
