### MergeSort.java 中的迭代归并

```java
public static void mergeSortIterative(int[] arr) {
    if (arr == null || arr.length <= 1) {
        return;
    }
    int n = arr.length;
    for (int step = 1; step < n; step *= 2) {
        for (int left = 0; left < n; left += 2 * step) {
            int mid = left + step - 1;
            if (mid >= n - 1) {
                break;
            }
            int right = Math.min(left + 2 * step - 1, n - 1);
            merge(arr, left, mid, right);
        }
    }
}
```

### QuickSort.java 中的随机快排

public static void quickSort2(int[] arr, int l, int r) {

## 5. 从朴素思路到优化思路

### 5.1 最直接的想法

最朴素就是冒泡、选择、插入排序。

### 5.2 为什么不够好

- 这些基础排序通常是 `O(n^2)`
- 数据规模大时不够快

### 5.3 优化方向

Java 原仓库里给出的优化方向就是更高阶排序：

- 归并排序：稳定，复杂度 `O(n log n)`
- 随机快排：平均 `O(n log n)`，常见且实战性强

这里先沿着原方案主讲迭代归并排序。


## 6. 核心算法讲解

### 6.1 算法名称

- 自底向上归并排序

### 6.2 为什么想到这个算法

因为归并排序天然能保证 `O(n log n)`，而迭代版又比较适合讲清楚整个排序过程。

### 6.3 关键状态

- `step`：当前有序小段长度
- `left`、`mid`、`right`：当前 merge 的左右边界
- `temp`：归并用辅助数组

### 6.4 步骤拆解

1. 把每个元素视作长度为 1 的有序段
2. 两两合并成长度为 2 的有序段
3. 再合并成长度为 4、8、16...
4. 直到整个数组有序


````python
class Solution:
    def sortArray(self, nums: list[int]) -> list[int]:
        if len(nums) <= 1:
            return nums

        n = len(nums)
        temp = [0] * n
        step = 1

        while step < n:
            left = 0
            while left < n:
                mid = left + step - 1
                if mid >= n - 1:
                    break
                right = min(left + 2 * step - 1, n - 1)
                self._merge(nums, temp, left, mid, right)
                left += 2 * step
            step *= 2
        return nums

    def _merge(self, arr: list[int], temp: list[int], l: int, m: int, r: int) -> None:
        i, j, k = l, m + 1, l
        while i <= m and j <= r:
            if arr[i] < arr[j]:
                temp[k] = arr[i]
                i += 1
            else:
                temp[k] = arr[j]
                j += 1
            k += 1
        while i <= m:
            temp[k] = arr[i]
            i += 1
            k += 1
        while j <= r:
            temp[k] = arr[j]
            j += 1
            k += 1
        for p in range(l, r + 1):
            arr[p] = temp[p]
````

## 7. 代码逐段讲解

### 7.1 为什么主讲归并排序

因为原 Java 入口把归并和快排都拿来验证了，但 `mergeSortIterative` 更适合 notebook 展示，而且代码结构完整独立。

### 7.2 `step` 的作用

`step` 表示当前每段有序区间的长度，每轮翻倍。

### 7.3 `_merge`

这部分和 Java 的 `merge` 方法一一对应：两个有序区间通过双指针合并回原数组。

### 7.4 原 Java 中的快排补充

原文件还调用了 `QuickSort.quickSort2`。那版快排用随机选 pivot + 荷兰国旗划分来降低退化风险，也是这题很值得掌握的另一条路线。


## 8. 复杂度分析

- 时间复杂度：`O(n log n)`
- 空间复杂度：`O(n)`
- 为什么是这个空间复杂度：需要辅助数组做 merge


## 9. 易错点

1. `step` 翻倍逻辑写错，导致有序段长度推进异常。
2. 最后一段不足 `step` 时边界处理出错。
3. merge 完后忘记把结果写回原数组。
4. 只看 `SortAnArray.java`，却忽略它真正依赖的是 `MergeSort.java` 和 `QuickSort.java` 里的核心实现。


## 10. 面试时怎么讲

可以这样概括：

> 这题原仓库里实际上同时验证了归并排序和随机快排两条方案。
> 如果我主讲一版，我会选自底向上的归并排序：先把数组看成很多长度为 1 的有序段，再不断两两合并。
> 每一轮有序段长度翻倍，直到整个数组有序。
> 时间复杂度是 `O(n log n)`，空间复杂度是 `O(n)`。
