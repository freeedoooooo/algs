## 1. 题目一句话总结

给定一个整数数组，要求你原地把它改造成“字典序刚好比当前更大一点”的那个排列；如果已经没有更大的排列了，就改成最小排列。

这题的核心不是暴力枚举，而是理解“下一个排列”在字典序上的结构规律。


## 2. 题目理解

### 2.1 题目要求

数组表示一个排列，我们要在 **原地** 修改它，使它变成：

- 比当前排列大
- 并且是所有更大排列里最小的那个

如果当前已经是最大排列，例如 `[3,2,1]`，那就返回最小排列 `[1,2,3]`。

### 2.2 关键约束

- 必须原地修改
- 不能额外复制所有排列
- 要求时间复杂度尽量低


## 3. 朴素思路

### 3.1 最直接的做法

先生成所有排列，排序后找到当前排列的下一个。

### 3.2 为什么不够好

- 全排列复杂度是 `O(n!)`
- 明显不满足面试和工程要求

### 3.3 关键观察

如果想让排列“刚好变大一点”，应该尽量：

1. 让高位尽量少变
2. 只在必须变的时候，改动最靠右的可提升位置
3. 低位改完以后要变成最小顺序

这就是这题的贪心本质。


## 4. 核心算法思路

### 4.1 算法名称

- 贪心
- 双指针
- 逆序区间处理

### 4.2 为什么想到这个算法

我们要求的是“下一个”排列，而不是“任意一个更大排列”。

这意味着：

- 应该从右往左找变化点
- 因为越靠右的位置变化，对整体字典序影响越小

### 4.3 关键步骤

1. 从右往左，找到第一个满足 `nums[i] < nums[i + 1]` 的位置 `i`
2. 如果找不到，说明整个数组是降序，直接整体反转
3. 如果找到了，再从右往左找第一个比 `nums[i]` 大的数 `nums[j]`
4. 交换 `nums[i]` 和 `nums[j]`
5. 将 `i + 1` 到末尾这一段反转，使其变成最小升序


## 5. 图解

### 5.1 先记住这道题到底在做什么

目标不是“变大就行”，而是：

- 先让排列变大
- 再保证这个变大的幅度尽可能小

所以它的动作一定是：

1. 从右边找一个必须修改的位置
2. 让这个位置只增大一点点
3. 把它右边重新排成最小

### 5.2 示例

```text
nums = [1, 2, 7, 4, 3, 1]
```

先标出索引，方便看“从右往左找拐点”的过程：

```text
索引:   0  1  2  3  4  5
数组: [ 1, 2, 7, 4, 3, 1 ]
```

从右往左看，相邻两项依次比较：

```text
3 > 1   继续往左
4 > 3   继续往左
7 > 4   继续往左
2 < 7   在这里停下
```

这说明：

- `i = 1`
- `nums[i] = 2`
- `i + 1` 到末尾这一段 `[7, 4, 3, 1]` 一定是降序
- `2` 是“最靠右的还能变大”的位置

也可以这样理解：右边这段已经大到不能再通过内部微调得到更大的下一个排列了，所以只能去提升左边第一个还能提升的位置。

### 5.3 找到刚刚好比它大的数

现在我们已经知道要改 `2`，但不能随便换成更大的数，而是要换成“刚刚好比它大”的那个。

因为后缀 `[7, 4, 3, 1]` 是降序，所以从右往左找第一个大于 `2` 的数，就是最合适的答案：

```text
[1, 2, 7, 4, 3, 1]
    ^        ^
    i        j
```

从右往左扫描：

```text
1 <= 2，不行
3 > 2，找到 j = 4
```

交换 `nums[i]` 和 `nums[j]`：

```text
[1, 3, 7, 4, 2, 1]
```

### 5.4 为什么交换后还要反转后缀

交换完以后，排列确实变大了，但还不是“下一个”排列，因为右侧后缀还太大：

```text
7, 4, 2, 1
```

为了让整体“只大一点点”，前面已经完成最小幅度提升，后面就必须尽量小。

而当前后缀本来就是降序，所以直接反转即可得到最小升序：

```text
[1, 3, 1, 2, 4, 7]
```

整个变化过程可以连起来看：

```text
[1, 2, 7, 4, 3, 1]
        找到拐点 i=1, nums[i]=2
              |
              v
[1, 3, 7, 4, 2, 1]
        交换 i 和 j
              |
              v
[1, 3, 1, 2, 4, 7]
        反转后缀，得到最终答案
```

### 5.5 特殊情况：整个数组本来就是降序

例如：

```text
[3, 2, 1]
```

从右往左永远找不到 `nums[i] < nums[i+1]`，说明它已经是字典序最大的排列。

这时候不存在“更大的下一个”，题目要求返回最小排列，所以直接整体反转：

```text
[1, 2, 3]
```

### 5.6 Mermaid 图解

```mermaid
flowchart TD
    A["原数组: [1,2,7,4,3,1]"] --> B["从右往左找第一个 nums[i] < nums[i+1]"]
    B --> C["找到 i=1, nums[i]=2"]
    C --> D["从右往左找第一个 > 2 的数"]
    D --> E["找到 j=4, nums[j]=3"]
    E --> F["交换 i 和 j: [1,3,7,4,2,1]"]
    F --> G["反转后缀 [7,4,2,1]"]
    G --> H["结果: [1,3,1,2,4,7]"]
```

图解要点：

- 从右往左找“最靠右的可提升位置”
- 右侧后缀在找到拐点时一定是降序
- 交换时要找“刚刚好更大”的元素
- 交换后，后缀只需要反转，不需要重新排序


````python
class Solution:
    def nextPermutation(self, nums: list[int]) -> None:
        n = len(nums)

        # 第一步：从右往左找第一个“下降转上升”的位置
        # 也就是找到第一个 nums[i] < nums[i + 1]
        i = n - 2
        while i >= 0 and nums[i] >= nums[i + 1]:
            i -= 1

        # 如果找到了，说明还有更大的排列
        if i >= 0:
            # 第二步：从右往左找第一个比 nums[i] 大的数
            # 因为右侧本身是降序，所以从右往左找到的第一个，
            # 就是“刚刚好大一点”的那个数
            j = n - 1
            while nums[j] <= nums[i]:
                j -= 1

            # 第三步：交换这两个位置
            nums[i], nums[j] = nums[j], nums[i]

        # 第四步：把 i 右边的后缀反转
        # 因为原后缀是降序，反转后就变成升序，也就是最小排列
        left, right = i + 1, n - 1
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1
````

## 6. 代码逐段讲解

### 6.1 找拐点

`i` 表示最靠右的、还能让排列变大的位置。

如果整个数组一直是降序，例如：

```text
[5,4,3,2,1]
```

那说明已经是最大排列，没有更大的了。

### 6.2 找替换值

右侧后缀本来是降序，所以从右往左找到第一个比 `nums[i]` 大的数，就是最合适的交换对象。

### 6.3 反转后缀

交换后，右侧仍然是降序或接近降序。

为了得到“紧邻的下一个排列”，必须让后缀尽可能小，所以直接反转成升序。


## 7. 复杂度分析

- 时间复杂度：`O(n)`
- 空间复杂度：`O(1)`

原因：

- 最多从右往左扫几次
- 再做一次区间反转
- 全部都是原地操作


## 8. 易错点

1. 找拐点时，循环条件容易写反。
2. 交换之后忘了反转后缀。
3. 误以为后缀要排序，其实反转就够了，因为它原本就是降序。
4. 没处理“整个数组已经是最大排列”的情况。


## 9. 面试时怎么讲

可以这样说：

> 这题要求的是字典序下一个排列，所以关键是尽量少改高位。  
> 我会从右往左找第一个还能变大的位置 `i`。  
> 然后再从右边找一个刚好比它大的数交换。  
> 最后把右侧后缀变成最小，也就是反转成升序。  
> 这样就能保证得到“刚刚比当前更大”的那个排列。  


## 10. 实际应用场景

### 10.1 对应什么现实问题

这题在现实里对应的是：

- 已经有一个当前方案
- 现在想要找到“下一个最接近的备选方案”
- 又不想把所有方案全部枚举出来

它本质上解决的是“**如何从当前状态高效推进到下一个合法候选状态**”的问题。

### 10.2 具体业务案例：智能排班系统中的备选顺序生成

#### 业务背景

先把场景简化：不是完整的复杂排班，而是“**固定岗位顺序下，员工如何占这些位置**”的问题。

例如今天只有 3 个岗位，并且岗位顺序固定不变：

- 第 1 个位置：前台
- 第 2 个位置：分拣
- 第 3 个位置：打包

员工有 3 个：`A`、`B`、`C`。

那么一个数组 `[A, B, C]` 的业务含义就不是单纯的字母排列，而是：

- `A` 占第 1 个位置，所以 A 去前台
- `B` 占第 2 个位置，所以 B 去分拣
- `C` 占第 3 个位置，所以 C 去打包

也就是说，在这个案例里：

- 数组下标 / 位置 = 固定岗位
- 数组元素 = 占这个岗位的人

这样一来，“一个排列”就和“一个岗位分配方案”直接对应上了。

#### 输入是什么

输入是一个表示当前岗位分配方案的数组，例如：

```text
[A, B, C]
```

它表示：

- 前台 = A
- 分拣 = B
- 打包 = C

#### 算法介入点

当前方案校验失败后，系统会调用“下一个排列”逻辑，直接生成字典序上的下一个候选顺序，然后再交给约束校验模块检查。

比如当前方案 `[A, B, C]` 不合法，原因可能是：

- B 今天不能做分拣
- 或者 A 今天不能在前台
- 或者这个人员组合不满足业务规则

这时系统并不是重新把所有方案都生成一遍，而是做：

1. 当前顺序是否合法
2. 如果不合法，调用 `next permutation` 生成下一个排列
3. 再次校验
4. 直到找到可用方案或结束

例如：

```text
[A, B, C] -> [A, C, B] -> [B, A, C] -> ...
```

每往后跳一次，就等于换了一个新的岗位分配方案。

#### 输出是什么

输出是一个新的候选顺序，例如：

```text
[A, C, B]
```

它对应的新排班就是：

- 前台 = A
- 分拣 = C
- 打包 = B

这个新方案会继续送到排班约束引擎中检查是否可用。

#### 它解决了什么问题

它解决的是：

> 已经有当前岗位分配方案时，如何快速得到“紧挨着的下一个候选方案”。

#### 如果不用这个算法，会有什么代价

这里要说准确一点：它并不是从本质上消灭枚举，而是让“从当前方案走到下一个方案”这一步更高效。

如果没有这个思路，系统可能会：

- 重新生成所有排列
- 对所有排列排序
- 再找到当前排列的下一个

而 `next permutation` 做的是：

- 你已经有当前排列
- 直接在 `O(n)` 时间里拿到下一个排列
- 不需要重建整个排列空间

所以它更准确的定位是：

> 它是一个高效的“下一个候选方案生成器”，不是完整的排班求解器。

### 10.3 常见疑问澄清

#### 疑问 1：`[A, B, C]` 为什么能表示排班方案？

因为岗位位置先固定了。

- 下标 0 = 前台
- 下标 1 = 分拣
- 下标 2 = 打包

所以元素放在哪个位置，就表示谁负责哪个岗位。

#### 疑问 2：这是不是还是暴力枚举？

是，从大方向上说，如果你要把所有候选方案一个个试过去，它本质上还是枚举。

但它不是“最笨的暴力枚举”，而是：

- 不重新生成所有排列
- 只根据当前排列，直接求下一个排列
- 用更低代价做有序推进

所以它属于“高效枚举器”，而不是“避免枚举的算法”。

#### 疑问 3：这个算法在排班里到底负责哪一步？

它只负责：

> 当前岗位分配失败后，生成下一个岗位分配候选方案。

它不负责：

- 判断这个方案是否满足工时规则
- 判断某人是否能上某个班
- 求最优排班结果

这些属于约束校验器、搜索器或求解器负责的事情。

### 10.4 在哪些技术或系统中有类似思想

- 约束求解器：从当前状态推进到下一个候选解
- 排班系统：生成下一个候选班次顺序
- 测试用例生成器：枚举下一个参数排列组合
- 推荐系统实验平台：尝试下一组排序优先级配置
- 搜索排序系统：在多个排序规则下生成相邻备选顺序

### 10.5 一句话总结

> 这道题真正解决的是：当当前岗位分配方案不可用时，如何不重建全部方案，而是直接跳到下一个最接近的候选方案。


````python
def next_permutation(nums: list[int]) -> list[int]:
    nums = nums[:]
    n = len(nums)
    i = n - 2

    while i >= 0 and nums[i] >= nums[i + 1]:
        i -= 1

    if i >= 0:
        j = n - 1
        while nums[j] <= nums[i]:
            j -= 1
        nums[i], nums[j] = nums[j], nums[i]

    left, right = i + 1, n - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1

    return nums


task_order = [1, 2, 7, 4, 3, 1]
print("当前顺序:", task_order)
print("下一个候选顺序:", next_permutation(task_order))
````

## 11. Demo 输出说明

- `当前顺序` 表示当前的任务执行排列
- `下一个候选顺序` 表示字典序刚好更大的下一种安排
