# p021 合并两个有序链表

- LeetCode: https://leetcode.cn/problems/merge-two-sorted-lists/
- 原始代码位置：`src/leetcode/p021_merge_two_sorted_lists/MergeTwoSortedLists.java`

这一题的目标是把两个递增链表合并成一个新的递增链表，并且尽量复用原有节点，不额外创建无关的新节点。


## 原代码思路整理

原 Java 文件给了两种写法：

1. 递归写法
2. 迭代写法

其中最核心的思路是一致的：

- 每一步都比较两个链表当前节点的值
- 把较小的那个节点接到结果链表后面
- 然后让对应链表继续向后走
- 如果某一边先走完，就把另一边剩余部分直接接上

## 补充说明

- 时间复杂度：`O(m + n)`
- 迭代写法额外空间：`O(1)`
- 递归写法额外空间：`O(m + n)` 的递归栈开销

在 LeetCode 提交时，通常更推荐迭代写法，因为它更稳，也不会受递归深度影响。


```python
from __future__ import annotations
from dataclasses import dataclass
from typing import Optional


@dataclass
class ListNode:
    val: int = 0
    next: Optional["ListNode"] = None


def build_linked_list(values: list[int]) -> Optional[ListNode]:
    dummy = ListNode()
    cur = dummy
    for value in values:
        cur.next = ListNode(value)
        cur = cur.next
    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> list[int]:
    ans = []
    while head:
        ans.append(head.val)
        head = head.next
    return ans
```

```python
class SolutionRecursive:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1:
            return list2
        if not list2:
            return list1

        # 谁更小，谁就作为当前这一轮的头节点
        if list1.val < list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
```

```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # 虚拟头结点，用来统一拼接逻辑
        dummy = ListNode()
        tail = dummy

        while list1 and list2:
            if list1.val <= list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next

        # 某一边已经走完，另一边剩余部分直接接上
        tail.next = list1 or list2
        return dummy.next
```

```python
list1 = build_linked_list([1, 2, 4])
list2 = build_linked_list([1, 3, 4])
result = Solution().mergeTwoLists(list1, list2)
linked_list_to_list(result)
```

```python
# 一边为空时，直接返回另一边
list1 = None
list2 = build_linked_list([0])
result = Solution().mergeTwoLists(list1, list2)
linked_list_to_list(result)
```
