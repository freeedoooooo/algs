# P002 两数相加

- LeetCode: https://leetcode.cn/problems/add-two-numbers/
- 原始代码位置：`src/leetcode/p002_add_two_numbers/AddTwoNumbers.java`

这一题的核心是模拟竖式加法。链表按逆序存储数字，所以我们可以从头到尾同步遍历两个链表，并维护一个进位 `carry`。


## 原代码思路整理

原 Java 注释里最关键的点可以整理成下面几条：

1. 如果某一位某个链表已经走完，就把它的值当成 `0`。
2. 使用虚拟头结点 `dummy`，可以避免单独处理头节点创建逻辑。
3. 每次计算当前位的和时，都要把上一位的进位一起算进去。
4. 全部遍历完成后，如果 `carry != 0`，还要额外补一个节点。

## 补充说明

- 时间复杂度：`O(max(m, n))`
- 额外辅助空间：`O(1)`（不计算返回结果链表）
- 这题最容易漏掉的是最后一位进位，比如 `999 + 1 = 1000`。


```python
from __future__ import annotations
from dataclasses import dataclass
from typing import Optional


@dataclass
class ListNode:
    val: int = 0
    next: Optional["ListNode"] = None


def build_linked_list(values: list[int]) -> Optional[ListNode]:
    dummy = ListNode()
    cur = dummy
    for value in values:
        cur.next = ListNode(value)
        cur = cur.next
    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> list[int]:
    ans = []
    while head:
        ans.append(head.val)
        head = head.next
    return ans
```

```python
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # 对应 Java 里的 head 虚拟头结点：
        # 用它统一处理结果链表的构造，避免单独讨论第一个节点怎么创建。
        dummy = ListNode()
        cur = dummy
        # carry 对应 Java 里的进位变量。
        carry = 0

        # 对应 Java 的 while (l1 != null || l2 != null) 逻辑，
        # 这里只是把“最后仍有进位”的情况也一并写进循环条件里。
        while l1 or l2 or carry:
            # 某一条链表如果已经走完，就把当前位当作 0。
            # 这和原 Java 里的三元表达式写法是一一对应的。
            value1 = l1.val if l1 else 0
            value2 = l2.val if l2 else 0

            # 当前位的总和 = 两个链表当前位 + 上一位进位。
            total = value1 + value2 + carry
            # divmod(total, 10) 就对应 Java 里的：
            # carry = sum / 10
            # value = sum % 10
            carry, digit = divmod(total, 10)

            # 创建当前位的新节点，并把结果链表尾指针向后移动。
            cur.next = ListNode(digit)
            cur = cur.next

            # 对应 Java 里的 if (l1 != null) l1 = l1.next。
            if l1:
                l1 = l1.next
            # 对应 Java 里的 if (l2 != null) l2 = l2.next。
            if l2:
                l2 = l2.next

        # 返回虚拟头结点的下一个节点，等价于 Java 里的 return head.next。
        return dummy.next
```

```python
# 示例 1：342 + 465 = 807
l1 = build_linked_list([2, 4, 3])
l2 = build_linked_list([5, 6, 4])
result = Solution().addTwoNumbers(l1, l2)
linked_list_to_list(result)
```

```python
# 边界示例：999 + 1 = 1000
l1 = build_linked_list([9, 9, 9])
l2 = build_linked_list([1])
result = Solution().addTwoNumbers(l1, l2)
linked_list_to_list(result)
```
