# p086 分隔链表

- LeetCode: https://leetcode.cn/problems/partition-list/
- 原始代码位置：`src/leetcode/p086_partition_list/PartitionList.java`

这一题要求把链表按给定值 `x` 分成两部分：所有小于 `x` 的节点放前面，所有大于等于 `x` 的节点放后面，并且要保持各自原有的相对顺序。


## 原代码思路整理

原 Java 解法的结构非常清晰：

1. 准备两条链表
   - 一条保存 `< x` 的节点
   - 一条保存 `>= x` 的节点
2. 遍历原链表，把每个节点摘下来，按值接到对应链表尾部
3. 最后把左边链表和右边链表拼起来

## 这题最关键的细节

原注释里特别强调了一点：

- 在处理当前节点时，要先保存 `next`
- 然后把当前节点的 `next` 断开

也就是：

```python
nxt = head.next
head.next = None
```

这样做的原因是：避免原链表残留的指针关系把后续节点错误地带进来，甚至在拼接后形成环。

## 复杂度

- 时间复杂度：`O(n)`
- 额外空间：`O(1)`


````python
from __future__ import annotations
from dataclasses import dataclass
from typing import Optional


@dataclass
class ListNode:
    val: int = 0
    next: Optional["ListNode"] = None


def build_linked_list(values: list[int]) -> Optional[ListNode]:
    dummy = ListNode()
    cur = dummy
    for value in values:
        cur.next = ListNode(value)
        cur = cur.next
    return dummy.next


def linked_list_to_list(head: Optional[ListNode]) -> list[int]:
    ans = []
    while head:
        ans.append(head.val)
        head = head.next
    return ans
````

````python
class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        # 对应 Java 里的 leftHead / leftTail：保存所有 < x 的节点。
        small_dummy = ListNode()
        # 对应 Java 里的 rightHead / rightTail：保存所有 >= x 的节点。
        large_dummy = ListNode()
        small_tail = small_dummy
        large_tail = large_dummy

        while head:
            # 对应 Java 里的：
            # ListNode next = head.next;
            # head.next = null;
            # 这样做是为了把当前节点从原链表里“摘干净”，避免后续拼接时串链或成环。
            nxt = head.next
            head.next = None

            # 对应 Java 的 if (head.val < x) ... else ... 分流逻辑。
            if head.val < x:
                # 把当前节点接到“小于 x”链表尾部。
                small_tail.next = head
                small_tail = small_tail.next
            else:
                # 把当前节点接到“大于等于 x”链表尾部。
                large_tail.next = head
                large_tail = large_tail.next

            # 继续处理原链表的下一个节点。
            head = nxt

        # 对应 Java 最后的 leftTail.next = rightHead。
        # 把两条子链表按顺序拼起来。
        small_tail.next = large_dummy.next
        # 返回新的头节点；如果左边为空，这里也会自然返回右边链表。
        return small_dummy.next
````

````python
head = build_linked_list([1, 4, 3, 2, 5, 2])
result = Solution().partition(head, 3)
linked_list_to_list(result)
````

````python
# 所有节点都小于 x，结果应保持原顺序
head = build_linked_list([1, 1, 2])
result = Solution().partition(head, 5)
linked_list_to_list(result)
````
