## 公共数据结构定义

链表题目中用到的节点类。

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Node:
    """带随机指针的链表节点（用于 P138）"""
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random

# 辅助函数：将列表转换为链表
def list_to_linked(lst):
    dummy = ListNode(0)
    cur = dummy
    for v in lst:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

# 辅助函数：将链表转换为列表（方便打印）
def linked_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result
```
