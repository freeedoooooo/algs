# p094 二叉树的中序遍历

- LeetCode: https://leetcode.cn/problems/binary-tree-inorder-traversal/
- 原始代码位置：`src/leetcode/p094_binary_tree_inorder_traversal/BinaryTreeInorderTraversal.java`

这一题要求按 `左 -> 根 -> 右` 的顺序返回二叉树节点值。原仓库里使用的是非递归写法，核心工具是栈。


## 原代码思路整理

原 Java 注释的逻辑可以整理成下面这套标准流程：

1. 先沿着当前节点一路向左走
2. 走到一个节点时，就把它压入栈中
3. 左边走到底以后，弹出栈顶节点并访问它
4. 然后转向这个节点的右子树，重复同样的过程

也可以把它理解成：

- 栈里保存的是“还没访问值，但左子树已经在处理路径上的节点”
- 每次弹栈，才真正把该节点的值加入答案

## 为什么要用栈

递归本质上也在利用调用栈保存遍历路径。非递归写法只是把这个过程显式地写出来，让我们自己管理入栈和出栈时机。

## 复杂度

- 时间复杂度：`O(n)`
- 空间复杂度：`O(h)`

其中 `h` 是树高；最坏情况下退化成链表时，空间复杂度为 `O(n)`。


```python
from __future__ import annotations
from dataclasses import dataclass
from typing import Optional


@dataclass
class TreeNode:
    val: int
    left: Optional["TreeNode"] = None
    right: Optional["TreeNode"] = None
```

```python
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> list[int]:
        # 对应 Java 里的 result 列表：保存中序遍历结果。
        result = []
        # 对应 Java 里的 Stack<TreeNode> stack。
        # 这个栈用来显式模拟递归调用路径。
        stack = []
        cur = root

        # 对应 Java 的 while (!stack.isEmpty() || head != null)。
        while stack or cur:
            # 第一步：一路向左，把沿途节点全部压栈。
            # 这和递归版不断进入 left 子树是同一个动作。
            while cur:
                stack.append(cur)
                cur = cur.left

            # 第二步：左边走到底后，弹出当前栈顶节点。
            # 这个节点此时的左子树已经处理完成，所以现在访问它的值。
            cur = stack.pop()
            result.append(cur.val)

            # 第三步：转向右子树。
            # 下一轮循环会继续对这棵右子树重复“先一路向左”的过程。
            cur = cur.right

        return result
```

```python
# 示例： [1, null, 2, 3]
#     1
#      \
#       2
#      /
#     3

root = TreeNode(1, None, TreeNode(2, TreeNode(3), None))
Solution().inorderTraversal(root)
```

```python
# 边界情况：空树
Solution().inorderTraversal(None)
```
